# String Manipulation in Python

Strings are one of the most used data types in Python. A **string** is a sequence of characters enclosed in quotes.

By the end of this notebook you will be able to:
- Create strings in various ways
- Index and slice strings
- Use the most important string methods
- Format strings with f-strings
- Understand string immutability
- Solve real text-processing problems

## 1. Creating Strings

Strings can use single quotes `'`, double quotes `"`, or triple quotes `'''`/`"""`.

```python
# Different ways to create strings
s1 = 'Hello'
s2 = "World"
s3 = '''This is
a multiline
string'''
s4 = "She said, 'hello'"  # single quotes inside double quotes
s5 = 'It\'s fine'         # escape the apostrophe
s6 = r'C:\Users\name'      # raw string — backslashes are literal

print(s1)
print(s2)
print(s3)
print(s4)
print(s5)
print(s6)
```

## 2. String Indexing and Slicing

Strings behave like lists of characters — you can index and slice them.

```python
text = 'Python'

# Indexing
print(text[0])   # 'P'
print(text[3])   # 'h'
print(text[-1])  # 'n' (last char)
print(text[-2])  # 'o'

# Slicing
print(text[0:3])   # 'Pyt'
print(text[2:])    # 'thon'
print(text[:4])    # 'Pyth'
print(text[::-1])  # 'nohtyP' — reversed!
```

## 3. String Immutability

Strings **cannot be changed** after creation. Operations return new strings.

```python
text = 'hello'

# You can't change a character in-place
try:
    text[0] = 'H'
except TypeError as e:
    print(f'Error: {e}')

# Instead, create a new string
text = 'H' + text[1:]  # reassign
print(text)  # 'Hello'
```

## 4. Essential String Methods

Python has dozens of string methods. Here are the most useful ones.

```python
# Case methods
text = 'Hello World'
print(text.upper())      # 'HELLO WORLD'
print(text.lower())      # 'hello world'
print(text.title())      # 'Hello World' (each word capitalized)
print(text.swapcase())   # 'hELLO wORLD'
print('hello'.capitalize()) # 'Hello'
```

```python
# Whitespace methods
text = '   hello world   '
print(repr(text.strip()))   # 'hello world' — removes both ends
print(repr(text.lstrip()))  # 'hello world   ' — left only
print(repr(text.rstrip()))  # '   hello world' — right only
```

```python
# Search methods
text = 'the quick brown fox jumps over the lazy dog'

print(text.find('fox'))      # 16 — first index, or -1 if not found
print(text.find('cat'))      # -1
print(text.count('the'))     # 2 — how many times
print(text.startswith('the')) # True
print(text.endswith('dog'))   # True
print('fox' in text)          # True — membership check
```

```python
# Split and join
sentence = 'apple,banana,cherry,date'

# Split — string → list
fruits = sentence.split(',')  # split on comma
print(fruits)  # ['apple', 'banana', 'cherry', 'date']

# Split on whitespace (default)
words = 'hello world python'.split()
print(words)  # ['hello', 'world', 'python']

# Join — list → string
joined = ' | '.join(fruits)
print(joined)  # 'apple | banana | cherry | date'

# Common join patterns
print(', '.join(['Alice', 'Bob', 'Charlie']))
print(''.join(['h', 'e', 'l', 'l', 'o']))
```

```python
# Replace
text = 'I love cats. Cats are great. My cat is fluffy.'

print(text.replace('cat', 'dog'))      # replaces all
print(text.replace('cat', 'dog', 1))   # replace only first occurrence
print(text.replace('Cats', 'Dogs'))    # case-sensitive!
```

## 5. String Formatting

Python has three ways to format strings. **f-strings** (Python 3.6+) are the modern recommended approach.

```python
name = 'Alice'
age = 30
score = 95.678

# 1. Old-style % formatting
print('Hello, %s! You are %d years old.' % (name, age))

# 2. str.format()
print('Hello, {}! You are {} years old.'.format(name, age))

# 3. f-strings (RECOMMENDED — Python 3.6+)
print(f'Hello, {name}! You are {age} years old.')

# f-strings can contain expressions
print(f'{name.upper()} is {age * 12} months old')
print(f'Score: {score:.2f}')  # 2 decimal places
print(f'{10 + 20}')           # expression
```

```python
# f-string formatting options
pi = 3.14159265
big_num = 1234567.89
fraction = 0.4567

print(f'{pi:.2f}')         # 3.14 — 2 decimal places
print(f'{pi:.5f}')         # 3.14159 — 5 decimal places
print(f'{big_num:,.2f}')   # 1,234,567.89 — thousands separator
print(f'{fraction:.1%}')   # 45.7% — percentage
print(f'{42:05d}')         # 00042 — zero-padded
print(f'{"hello":>10}')    # right-align in width 10
print(f'{"hello":<10}|')   # left-align in width 10
print(f'{"hello":^10}')    # center in width 10
```

## 6. Checking String Content

```python
print('hello123'.isalnum())  # True — letters and digits only
print('hello'.isalpha())     # True — letters only
print('12345'.isdigit())     # True — digits only
print('   '.isspace())       # True — whitespace only
print('Hello World'.istitle()) # True — title case
print('hello'.islower())      # True
print('HELLO'.isupper())      # True
```

## 7. String Concatenation and Repetition

```python
# Concatenation with +
first = 'Hello'
last = 'World'
full = first + ', ' + last + '!'
print(full)

# Repetition with *
print('ha' * 3)      # 'hahaha'
print('-' * 30)      # a line of 30 dashes

# Building strings efficiently with join (not +)
parts = ['one', 'two', 'three', 'four']
result = ' '.join(parts)  # much faster than looping with +
print(result)
```

## Common Mistakes

### Mistake 1: Concatenating String and Number Directly

```python
age = 25

# BAD
try:
    print('I am ' + age + ' years old')  # TypeError!
except TypeError as e:
    print(f'Error: {e}')

# GOOD
print('I am ' + str(age) + ' years old')  # convert to string
print(f'I am {age} years old')            # f-string (best)
```

### Mistake 2: Using `==` with Different Cases

```python
user_input = 'YES'

# BAD — case-sensitive comparison
if user_input == 'yes':
    print('User agreed')  # won't print!

# GOOD — normalize case first
if user_input.lower() == 'yes':
    print('User agreed')  # works!
```

### Mistake 3: Building Strings with `+` in a Loop

```python
words = ['the', 'quick', 'brown', 'fox']

# BAD — creates a new string object each iteration (slow for large lists)
result = ''
for word in words:
    result += word + ' '
print(result.strip())

# GOOD — collect parts, join at the end
result = ' '.join(words)
print(result)
```

### Exercise 2 (Medium): Palindrome Checker
Write a function `is_palindrome(s)` that returns True if the string reads the same forwards and backwards. Ignore spaces and case.

```python
# Your solution here
```

```python
# Solution
def is_palindrome(s):
    cleaned = s.lower().replace(' ', '')
    return cleaned == cleaned[::-1]

print(is_palindrome('racecar'))           # True
print(is_palindrome('A man a plan a canal Panama'))  # True
print(is_palindrome('hello'))             # False
```

### Exercise 3 (Hard): Caesar Cipher
Implement a Caesar cipher: shift each letter by `n` positions (wrapping around). Non-letters stay unchanged.

```python
# Your solution here
```

```python
# Solution
def caesar_cipher(text, shift):
    result = []
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shifted = (ord(char) - base + shift) % 26 + base
            result.append(chr(shifted))
        else:
            result.append(char)  # keep non-letters unchanged
    return ''.join(result)

print(caesar_cipher('Hello, World!', 3))   # 'Khoor, Zruog!'
print(caesar_cipher('Khoor, Zruog!', -3))  # 'Hello, World!' (decode)
```

## Mini-Project: Text Analyzer

Build a text analyzer that reports word count, character count, most common words, and other stats.

```python
def analyze_text(text):
    """Analyze text and print a detailed report."""
    # Basic counts
    char_count = len(text)
    char_no_space = len(text.replace(' ', ''))
    
    words = text.lower().split()
    word_count = len(words)
    
    sentences = text.split('.')
    sentence_count = len([s for s in sentences if s.strip()])
    
    # Unique words
    unique_words = set(words)
    
    # Word frequency
    freq = {}
    for word in words:
        # Remove punctuation from word
        clean = ''.join(c for c in word if c.isalpha())
        if clean:
            freq[clean] = freq.get(clean, 0) + 1
    
    # Top 5 most common words
    top_words = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:5]
    
    # Print report
    print('=' * 40)
    print('TEXT ANALYSIS REPORT')
    print('=' * 40)
    print(f'Characters (total):    {char_count}')
    print(f'Characters (no spaces): {char_no_space}')
    print(f'Words:                  {word_count}')
    print(f'Unique words:           {len(unique_words)}')
    print(f'Sentences:              {sentence_count}')
    print(f'Avg word length:        {sum(len(w) for w in words)/word_count:.1f}')
    print()
    print('Top 5 most common words:')
    for word, count in top_words:
        bar = '#' * count
        print(f'  {word:15} {count:3}  {bar}')
    print('=' * 40)


sample = """
Python is a great programming language. Python is easy to learn.
Python is used in web development, data science, and automation.
Many great programmers use Python every day.
"""

analyze_text(sample)
```
