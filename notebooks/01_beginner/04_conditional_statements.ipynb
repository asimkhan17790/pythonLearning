## 1. if / elif / else

The `if` statement is like a fork in the road. Python checks the condition:
- If it's `True` → run the indented block
- If it's `False` → skip it

`elif` ("else if") adds more forks. `else` is the final catch-all — it runs only when ALL previous conditions were `False`.

**Critical rule: Indentation defines the block.** Python uses 4 spaces (or 1 tab) to know which lines belong to which condition.

````python
# Simple if statement
temperature = 35

if temperature > 30:
    print('It is hot today!')        # only runs if condition is True
    print('Stay hydrated.')          # also part of the if block (same indent)

print('This always runs.')           # outside the if block (no extra indent)
````

````python
# if / else — two options
hour = 14   # 2 PM in 24-hour format

if hour < 12:
    greeting = 'Good morning!'
else:
    greeting = 'Good afternoon/evening!'

print(greeting)
````

````python
# if / elif / else — multiple options
# Only ONE branch runs — the first one whose condition is True.

score = 83

if score >= 90:
    grade = 'A'       # not reached — 83 < 90
elif score >= 80:
    grade = 'B'       # this runs — 83 >= 80
elif score >= 70:
    grade = 'C'       # skipped — already matched above
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f'Score {score} → Grade {grade}')
````

````python
# Real-world example: shipping cost calculator
order_weight_kg = 4.5
destination     = 'international'

if destination == 'local' and order_weight_kg <= 1:
    shipping = 150
    label    = 'Local Standard'
elif destination == 'local' and order_weight_kg <= 5:
    shipping = 300
    label    = 'Local Bulk'
elif destination == 'national':
    shipping = 500 + (order_weight_kg * 50)
    label    = 'National'
elif destination == 'international':
    shipping = 2000 + (order_weight_kg * 300)
    label    = 'International'
else:
    shipping = 0
    label    = 'Unknown destination'

print(f'Order: {order_weight_kg}kg → {label} shipping')
print(f'Shipping cost: PKR {shipping:,.0f}')
````

## 2. Ternary Expression (One-line if)

Python lets you write a simple if/else on a **single line**. This is called a **ternary expression** or **conditional expression**.

```python
value = result_if_true  if  condition  else  result_if_false
```

Use it for **simple, short decisions** where readability is maintained. Avoid nesting it — that gets confusing fast.

````python
# Basic ternary
age = 20
status = 'adult' if age >= 18 else 'minor'
print(f'Status: {status}')

# Compare to the multi-line version:
if age >= 18:
    status = 'adult'
else:
    status = 'minor'
# Both are identical in behaviour — choose based on readability
````

````python
# Ternary inside f-strings — very common pattern
marks   = [78, 45, 92, 38, 88]

for i, mark in enumerate(marks):
    result = 'Pass' if mark >= 50 else 'Fail'
    status_icon = '✓' if mark >= 50 else '✗'
    print(f'  Student {i+1}: {mark:3d}  {status_icon}  {result}')

# Ternary for selecting a value
discount_rate = 0.15 if len(marks) >= 5 else 0.10
print(f'\nBulk discount rate: {discount_rate:.0%}')
````

````python
# What NOT to do — nested ternaries are hard to read
score = 75

# BAD — avoid this:
grade = 'A' if score >= 90 else 'B' if score >= 80 else 'C' if score >= 70 else 'F'
print('Nested ternary (hard to read):', grade)

# GOOD — use if/elif/else for multiple branches:
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
print('Clear if/elif (easy to read):', grade)
````

## 3. Nested Conditions

You can put an `if` statement **inside another `if` statement**. This is called **nesting**. Each level of nesting gets another 4 spaces of indentation.

Be careful: too many levels of nesting makes code hard to read. If you have more than 2-3 levels, consider restructuring.

````python
# Nested conditions example: loan eligibility
age         = 28
income      = 75_000
credit_score = 720
has_existing_loan = False

print('Loan Application Check:')

if age >= 21:                              # first check
    print('  Age: OK')
    if income >= 50_000:                   # second check (nested)
        print('  Income: OK')
        if credit_score >= 700:            # third check (double nested)
            print('  Credit Score: OK')
            if not has_existing_loan:      # fourth check
                print('  No existing loan: OK')
                print('\nResult: APPROVED for loan!')
            else:
                print('  Rejected: Existing loan detected')
        else:
            print(f'  Rejected: Credit score {credit_score} too low (need 700+)')
    else:
        print(f'  Rejected: Income PKR {income:,} too low (need 50,000+)')
else:
    print(f'  Rejected: Age {age} too young (must be 21+)')
````

````python
# Flattening nested conditions with 'and' — often cleaner
# The same loan logic, but without deep nesting:

eligible = (
    age >= 21
    and income >= 50_000
    and credit_score >= 700
    and not has_existing_loan
)

if eligible:
    print('APPROVED for loan!')
else:
    # Now we check individually to give a reason
    reasons = []
    if age < 21:              reasons.append('must be 21+')
    if income < 50_000:       reasons.append('income too low')
    if credit_score < 700:    reasons.append('credit score too low')
    if has_existing_loan:     reasons.append('has existing loan')
    print(f'REJECTED: {', '.join(reasons)}')
````

## 4. Truthiness — What Python Considers False

In Python, **any value** can be used as a condition in an `if` statement. Python automatically converts it to `True` or `False`. This is called "truthiness".

**Values that are FALSY (treated as False):**
- `False` — the boolean itself
- `0`, `0.0` — any zero number
- `''`, `""` — empty string
- `[]` — empty list
- `{}` — empty dictionary
- `()` — empty tuple
- `set()` — empty set
- `None`

**Everything else is TRUTHY.**

````python
# Falsy values — all treated as False in an if statement
falsy_values = [0, 0.0, '', [], {}, (), None, False]

for val in falsy_values:
    if val:
        print(f'{repr(val):<10} → Truthy')
    else:
        print(f'{repr(val):<10} → Falsy')
````

````python
# Practical use: checking if a value exists
username = ''   # empty — user didn't fill in the field

# Instead of: if username != '':
if not username:   # clean Pythonic way to check for empty/None
    print('Username is required!')

username = 'Ali'
if username:       # non-empty string is truthy
    print(f'Welcome, {username}!')
````

````python
# Checking empty collections
notifications = []    # empty list — no notifications

if notifications:     # empty list is falsy
    print(f'You have {len(notifications)} notifications')
else:
    print('No notifications')

notifications = ['Friend request', 'New message', 'Like on post']
if notifications:
    print(f'You have {len(notifications)} notifications:')
    for n in notifications:
        print(f'  - {n}')
````

## 5. Short-Circuit Evaluation in Conditions

Python uses **short-circuit evaluation** when evaluating `and`/`or` chains. It stops as soon as the result is determined.

This isn't just an optimisation — it's a safety feature: you can check if a variable exists **before** using it.

````python
# Safe attribute access pattern
# Check the container first, then access inside it

user_data = {'name': 'Sana', 'age': 25}   # imagine this could be None

# Safe: if user_data is None, the second condition is never evaluated
if user_data and user_data.get('age', 0) >= 18:
    print(f'{user_data["name"]} is an adult')

# Without short-circuit, this would crash when user_data is None:
user_data = None
if user_data and user_data.get('age', 0) >= 18:  # safe because of short-circuit!
    print('Adult')
else:
    print('No user data or underage')
````

````python
# 'or' as a default value pattern
# 'x or default' returns x if x is truthy, else returns default

username = ''
display_name = username or 'Anonymous'    # if username is empty, use 'Anonymous'
print(f'Hello, {display_name}!')

username = 'Tariq'
display_name = username or 'Anonymous'
print(f'Hello, {display_name}!')

# 'and' as a conditional return pattern
# 'x and y' returns y if x is truthy, else returns x
items = [1, 2, 3]
first = items and items[0]    # only access items[0] if items is not empty
print(f'First item: {first}')
````

## 6. match / case (Python 3.10+)

`match`/`case` is Python's version of a **switch statement** (found in C, Java, etc.). It compares a single value against multiple patterns.

It's most useful when you have many exact value comparisons — cleaner than a long `if/elif` chain.

````python
# Basic match/case
day_number = 3

match day_number:
    case 1:
        day_name = 'Monday'
    case 2:
        day_name = 'Tuesday'
    case 3:
        day_name = 'Wednesday'
    case 4:
        day_name = 'Thursday'
    case 5:
        day_name = 'Friday'
    case 6 | 7:            # the | means 'or' — matches 6 OR 7
        day_name = 'Weekend'
    case _:                # _ is the wildcard — matches anything else
        day_name = 'Invalid day'

print(f'Day {day_number} is: {day_name}')
````

````python
# match/case with string commands
command = 'start'

match command.lower():
    case 'start' | 'begin' | 'go':
        print('Starting the process...')
    case 'stop' | 'end' | 'quit' | 'exit':
        print('Stopping the process...')
    case 'help':
        print('Available commands: start, stop, status, help')
    case 'status':
        print('System is running normally')
    case _:
        print(f'Unknown command: "{command}". Type "help" for options.')
````

````python
# match/case with HTTP status codes — very practical!
def describe_status(status_code):
    match status_code:
        case 200:
            return 'OK — Request successful'
        case 201:
            return 'Created — Resource created successfully'
        case 301 | 302:
            return 'Redirect — Page has moved'
        case 400:
            return 'Bad Request — Invalid input'
        case 401:
            return 'Unauthorized — Login required'
        case 403:
            return 'Forbidden — Access denied'
        case 404:
            return 'Not Found — Page does not exist'
        case 500:
            return 'Internal Server Error — Server problem'
        case _:
            return f'Unknown status code: {status_code}'

for code in [200, 404, 401, 500, 999]:
    print(f'{code}: {describe_status(code)}')
````

## 7. Common Mistakes

````python
# MISTAKE 1: Forgetting that elif/else must align with if
# Python is strict about indentation — misalignment causes errors or wrong results.

x = 5

# WRONG (each if is independent — all three might run!):
if x > 0:
    print('Positive')
if x > 3:          # This is a SEPARATE if, not elif!
    print('Greater than 3')

print()

# CORRECT (elif — only ONE branch runs):
if x < 0:
    print('Negative')
elif x == 0:
    print('Zero')
elif x > 0:
    print('Positive')   # Only this runs
````

````python
# MISTAKE 2: Comparing to True/False explicitly (redundant)
is_active = True

# Redundant — don't do this:
if is_active == True:
    print('Active (verbose)')

# Pythonic — just use the boolean directly:
if is_active:
    print('Active (clean)')

# Same for False:
if not is_active:   # not: if is_active == False:
    print('Not active')

# Also applies to None:
value = None
if value is None:    # correct
    print('No value')
````

````python
# MISTAKE 3: Range check order matters with elif
score = 85

# WRONG ORDER — the first condition is too broad!
if score >= 60:      # True for 85 — this catches EVERYTHING above 60
    print('D or above')  # All scores 60+ land here — wrong!
elif score >= 80:
    print('B or above')  # Never reached!

print()

# CORRECT ORDER — most specific (highest) first:
if score >= 90:
    print('A')
elif score >= 80:    # only reaches here if score < 90
    print('B')       # This correctly runs for score=85
elif score >= 70:
    print('C')
elif score >= 60:
    print('D')
else:
    print('F')
````

## 9. Solutions

````python
# SOLUTION 1 - try it yourself first!
num = int(input('Enter a number: '))

if num > 0:
    print('The number is positive.')
    print('It is', 'even.' if num % 2 == 0 else 'odd.')
elif num < 0:
    print('The number is negative.')
    print('It is', 'even.' if num % 2 == 0 else 'odd.')
else:
    print('The number is zero.')
````

````python
# SOLUTION 2 - try it yourself first!
age        = int(input('Enter visitor age: '))
is_weekend = input('Is it a weekend? (yes/no): ').lower() == 'yes'

if age < 5:
    ticket_type  = 'Child (Free)'
    base_price   = 0
elif age <= 12:
    ticket_type  = 'Child'
    base_price   = 500
elif age <= 17:
    ticket_type  = 'Teenager'
    base_price   = 800
elif age <= 59:
    ticket_type  = 'Adult'
    base_price   = 1200
else:
    ticket_type  = 'Senior'
    base_price   = 600

surcharge    = base_price * 0.20 if is_weekend else 0
final_price  = base_price + surcharge

print(f'\nTicket type: {ticket_type}')
print(f'Base price:  PKR {base_price}')
if is_weekend:
    print(f'Weekend surcharge (20%): PKR {surcharge:.0f}')
print(f'Final price: PKR {final_price:.0f}')
````

````python
# SOLUTION 3 - try it yourself first!
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for n in range(1, 31):
    if n % 3 == 0 and n % 5 == 0:
        print(f'{n:3d}: FizzBuzz')
    elif n % 3 == 0:
        print(f'{n:3d}: Fizz')
    elif n % 5 == 0:
        print(f'{n:3d}: Buzz')
    elif is_prime(n):
        print(f'{n:3d}: PrimeBuzz')
    else:
        print(f'{n:3d}: {n}')
````

## 10. Mini-Project: Simple ATM

````python
# ============================================================
# MINI-PROJECT: Simple ATM Simulator
# ============================================================
# Simulates an ATM: check balance, withdraw, deposit.
# Demonstrates nested conditions and thorough edge case handling.

# --- Account setup ---
account_holder = 'Ayesha Farooq'
balance        = 25_000.00   # PKR
pin            = '4729'
daily_limit    = 50_000.00
withdrawn_today = 0.0

# --- ATM interface ---
print('=' * 40)
print(f'{"NATIONAL BANK ATM":^40}')
print('=' * 40)

# PIN verification
entered_pin = input('Enter your PIN: ')

if entered_pin != pin:
    print('\nIncorrect PIN. Transaction cancelled.')
else:
    print(f'\nWelcome, {account_holder}!')
    print('\nSelect transaction:')
    print('  1. Check Balance')
    print('  2. Withdraw')
    print('  3. Deposit')

    choice = input('\nEnter choice (1/2/3): ').strip()

    if choice == '1':
        # Check balance
        print(f'\nAvailable balance: PKR {balance:,.2f}')

    elif choice == '2':
        # Withdraw
        raw_amount = input('Enter withdrawal amount: PKR ')

        # Validate that input is a number
        try:
            amount = float(raw_amount)
        except ValueError:
            print('Invalid amount entered.')
            amount = -1   # sentinel to skip the rest

        if amount > 0:
            if amount > balance:
                print(f'Insufficient funds. Balance: PKR {balance:,.2f}')
            elif amount > daily_limit - withdrawn_today:
                remaining_limit = daily_limit - withdrawn_today
                print(f'Daily limit exceeded. Remaining today: PKR {remaining_limit:,.2f}')
            elif amount % 500 != 0:
                print('ATM dispenses in multiples of PKR 500 only.')
            else:
                balance -= amount
                withdrawn_today += amount
                print(f'\nPlease collect PKR {amount:,.0f}')
                print(f'Remaining balance: PKR {balance:,.2f}')
        elif amount <= 0 and amount != -1:
            print('Amount must be positive.')

    elif choice == '3':
        # Deposit
        raw_amount = input('Enter deposit amount: PKR ')
        try:
            amount = float(raw_amount)
            if amount <= 0:
                print('Deposit amount must be positive.')
            elif amount > 500_000:
                print('Maximum single deposit is PKR 500,000.')
            else:
                balance += amount
                print(f'\nPKR {amount:,.2f} deposited successfully.')
                print(f'New balance: PKR {balance:,.2f}')
        except ValueError:
            print('Invalid amount entered.')

    else:
        print(f'Invalid choice: "{choice}". Please select 1, 2, or 3.')

    print('\nThank you for using National Bank ATM!')

print('=' * 40)
````
