## 1. Arithmetic Operators

These operators perform mathematical calculations. Most are familiar from maths class — the new ones to learn are `//` (floor division), `%` (modulo), and `**` (power).

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `7 + 3` | `10` |
| `-` | Subtraction | `7 - 3` | `4` |
| `*` | Multiplication | `7 * 3` | `21` |
| `/` | Division | `7 / 3` | `2.333...` |
| `//` | Floor Division | `7 // 3` | `2` |
| `%` | Modulo (remainder) | `7 % 3` | `1` |
| `**` | Exponentiation | `7 ** 3` | `343` |

```python
# Basic arithmetic
students  = 35
new_batch = 12

total     = students + new_batch    # 47
remaining = students - 5            # 30 (5 graduated)
doubled   = students * 2            # 70
share     = students / 5            # 7.0  (always float with /)

print(f'Total after new batch: {total}')
print(f'After 5 graduate: {remaining}')
print(f'Share per group: {share}')
```

```python
# Floor division (//) and Modulo (%) — the inseparable pair
# Floor division: how many WHOLE times does divisor fit?
# Modulo: what is LEFT OVER after floor division?

total_minutes = 247

hours   = total_minutes // 60   # 4  (4 complete hours)
minutes = total_minutes % 60    # 7  (7 minutes left over)

print(f'{total_minutes} minutes = {hours} hours and {minutes} minutes')

# Modulo practical uses:
number = 18
print(f'Is {number} even?', number % 2 == 0)   # True if remainder is 0
print(f'Is {number} divisible by 3?', number % 3 == 0)

# Cycling through a fixed range (very useful in games and animations)
for i in range(7):
    day_index = i % 5   # repeats 0,1,2,3,4,0,1,...
    days = ['Mon','Tue','Wed','Thu','Fri']
    print(f'Day {i}: {days[day_index]}')
```

```python
# Exponentiation (**) — raising to a power
side   = 5
area   = side ** 2          # 25  (square)
volume = side ** 3          # 125 (cube)
root   = 144 ** 0.5         # 12.0 (square root via power of 0.5)
cube_root = 27 ** (1/3)     # 3.0

print(f'Side: {side}')
print(f'Area (side²): {area}')
print(f'Volume (side³): {volume}')
print(f'Square root of 144: {root}')
print(f'Cube root of 27: {cube_root:.1f}')

# Compound interest formula: A = P(1 + r/n)^(nt)
principal  = 100_000    # PKR
rate       = 0.12       # 12% annual
years      = 5
amount     = principal * (1 + rate) ** years
print(f'\nPKR {principal:,} at 12% for {years} years = PKR {amount:,.2f}')
```

## 2. Comparison Operators

Comparison operators compare two values and always return `True` or `False`.

| Operator | Meaning | Example |
|----------|---------|--------|
| `==` | Equal to | `5 == 5` → `True` |
| `!=` | Not equal to | `5 != 3` → `True` |
| `>` | Greater than | `7 > 3` → `True` |
| `<` | Less than | `7 < 3` → `False` |
| `>=` | Greater than or equal | `5 >= 5` → `True` |
| `<=` | Less than or equal | `4 <= 5` → `True` |

```python
# Comparing numbers
my_score    = 78
pass_mark   = 60
top_score   = 95

print(f'Did I pass? {my_score >= pass_mark}')         # True
print(f'Did I top? {my_score == top_score}')          # False
print(f'Below top? {my_score < top_score}')           # True
print(f'Score gap from top: {top_score - my_score}')  # 17
```

```python
# Comparing strings — alphabetical (lexicographic) order
print('apple' == 'apple')   # True
print('apple' == 'Apple')   # False — case matters!
print('apple' < 'banana')   # True — 'a' comes before 'b'
print('cat' > 'car')        # True — 't' comes after 'r'

# Chained comparisons — very Pythonic!
age = 25
print(18 <= age <= 65)      # True — Python allows chaining!
# Equivalent to: (18 <= age) and (age <= 65)

temperature = 23
comfortable = 18 <= temperature <= 28
print(f'Temperature {temperature}°C is comfortable: {comfortable}')
```

## 3. Logical Operators

Logical operators combine multiple conditions into one.

- `and` — True only if **both** sides are True
- `or` — True if **at least one** side is True
- `not` — **reverses** the boolean value

Think of `and` as a strict bouncer (both requirements must be met) and `or` as a lenient bouncer (either one is fine).

```python
# and — both must be True
has_ticket = True
has_id     = True
is_adult   = False

can_enter_concert = has_ticket and has_id
print(f'Can enter concert: {can_enter_concert}')   # True

can_buy_alcohol = has_id and is_adult
print(f'Can buy alcohol: {can_buy_alcohol}')        # False

# or — at least one must be True
pays_cash  = False
pays_card  = True
can_pay    = pays_cash or pays_card
print(f'Can pay: {can_pay}')   # True

# not — flip True to False and vice versa
is_closed  = False
is_open    = not is_closed
print(f'Shop is open: {is_open}')   # True
```

```python
# Truth table for and/or
print('--- AND truth table ---')
for a in [True, False]:
    for b in [True, False]:
        print(f'{str(a):<6} and {str(b):<6} = {a and b}')

print()
print('--- OR truth table ---')
for a in [True, False]:
    for b in [True, False]:
        print(f'{str(a):<6} or  {str(b):<6} = {a or b}')
```

```python
# Short-circuit evaluation — Python is lazy about evaluating
# With 'and': if the left side is False, Python skips the right side
# With 'or':  if the left side is True, Python skips the right side

def expensive_check():
    print('  (expensive check ran!)')   # shows if this function was called
    return True

is_active = False
print('Testing: False and expensive_check()')
result = is_active and expensive_check()  # expensive_check is NEVER called!
print(f'Result: {result}')               # False — short-circuited

print()
is_active = True
print('Testing: True or expensive_check()')
result = is_active or expensive_check()   # expensive_check is NEVER called!
print(f'Result: {result}')                # True — short-circuited
```

## 4. Assignment Shortcut Operators

These operators are shortcuts: `x += 5` means `x = x + 5`. They modify a variable *in place*, making code shorter and more readable.

```python
# Without shortcuts (verbose)
score = 100
score = score + 50   # add a bonus
score = score - 10   # deduct a penalty
print(score)         # 140

# With shortcuts (cleaner)
score = 100
score += 50     # equivalent to: score = score + 50
score -= 10     # equivalent to: score = score - 10
print(score)    # 140

# All shortcut operators:
x = 20
x += 5;   print(f'After +=5:   {x}')   # 25
x -= 3;   print(f'After -=3:   {x}')   # 22
x *= 2;   print(f'After *=2:   {x}')   # 44
x //= 4;  print(f'After //=4:  {x}')   # 11
x **= 2;  print(f'After **=2:  {x}')   # 121
x %= 100; print(f'After %=100: {x}')   # 21
```

```python
# Practical example: running total
cart_total = 0

prices = [450, 1200, 85, 320, 999]
for price in prices:
    cart_total += price     # accumulate the total

print(f'Cart total: PKR {cart_total:,}')

# Apply a 10% discount
cart_total *= 0.90
print(f'After 10% discount: PKR {cart_total:,.2f}')

# String shortcuts work too!
report = 'Monthly Report: '
report += 'March 2024'
print(report)
```

## 5. Identity Operators — `is` and `is not`

`is` checks if two variables point to the **exact same object in memory**, not just if they have equal values.

Think of it this way: two identical cars (`==`) are equal, but `is` asks "is this literally the SAME car?"

**Rule of thumb:** Use `is` only to compare against `None`, `True`, or `False`. Use `==` for value comparisons.

```python
# The classic use case: checking for None
result = None

# Correct way:
if result is None:
    print('No result yet')   # this prints

# Also valid but less idiomatic:
if result == None:
    print('No result (== version)')

# is not None — checking that something exists
data = 'some data'
if data is not None:
    print(f'Data found: {data}')
```

```python
# Why == and 'is' are different for objects
list1 = [1, 2, 3]
list2 = [1, 2, 3]   # same values, but a DIFFERENT object
list3 = list1       # same object (list3 is an alias for list1)

print(list1 == list2)    # True  — same VALUES
print(list1 is list2)    # False — different OBJECTS
print(list1 is list3)    # True  — same OBJECT

# For small integers and strings, Python caches them — so 'is' may give True
# but NEVER rely on this behaviour. Always use == for value comparison.
a = 5
b = 5
print(a is b)   # True (due to Python's integer caching) — but don't count on it!
```

## 6. Membership Operators — `in` and `not in`

`in` checks whether a value exists **inside** a sequence (string, list, tuple, etc.). It's one of the most naturally readable features of Python.

```python
# Membership in strings
email = 'user@example.com'
print('@' in email)         # True — valid email structure
print('.com' in email)      # True
print('gmail' in email)     # False

# Membership in lists
approved_countries = ['Pakistan', 'Turkey', 'UAE', 'Malaysia', 'Egypt']
user_country = 'Turkey'

if user_country in approved_countries:
    print(f'{user_country} is approved for travel')

banned_words = ['spam', 'scam', 'fraud', 'hack']
message = 'Buy now for a great discount!'
is_spam = any(word in message.lower() for word in banned_words)
print(f'Message flagged as spam: {is_spam}')
```

```python
# not in — check absence
vowels = 'aeiou'
letter = 'k'

if letter not in vowels:
    print(f'\'{letter}\' is a consonant')

# Practical: valid choice checking
valid_sizes = ['S', 'M', 'L', 'XL', 'XXL']
chosen_size = 'XS'

if chosen_size not in valid_sizes:
    print(f'Error: size "{chosen_size}" is not available.')
    print(f'Available sizes: {', '.join(valid_sizes)}')
```

## 7. Operator Precedence

When an expression has multiple operators, Python uses a fixed order to evaluate them — just like BODMAS/PEMDAS in maths.

**Highest to lowest priority:**
1. `()` — parentheses (highest — always evaluated first)
2. `**` — exponentiation
3. `+x`, `-x` — unary plus/minus
4. `*`, `/`, `//`, `%` — multiplication and division
5. `+`, `-` — addition and subtraction
6. `==`, `!=`, `<`, `>`, `<=`, `>=`, `is`, `in` — comparisons
7. `not` — logical not
8. `and` — logical and
9. `or` — logical or (lowest)

```python
# Arithmetic precedence
print(2 + 3 * 4)       # 14, not 20 (multiplication first)
print((2 + 3) * 4)     # 20 (parentheses override)
print(2 ** 3 ** 2)     # 512 (** is right-associative: 2**(3**2) = 2**9 = 512)
print((2 ** 3) ** 2)   # 64  (left to right when parenthesised)
print(10 - 2 * 3 + 1)  # 5   (multiply first, then left-to-right)

# Mixed with comparisons
x = 10
print(x + 5 > 12)         # True  (x+5=15, 15>12=True)
print(not x > 5 and True) # False (x>5=True, not True=False, False and True=False)
```

```python
# Real-world example: tax calculation
salary        = 80_000
bonus         = 15_000
tax_rate      = 0.10
transport_allowance = 5_000

# Without parentheses — is this right?
tax_wrong  = salary + bonus * tax_rate  # salary + (bonus * tax_rate) — probably wrong!

# With parentheses — clear intent
tax_correct = (salary + bonus) * tax_rate

net_salary = salary + bonus - tax_correct + transport_allowance

print(f'Gross: PKR {salary + bonus:,}')
print(f'Tax (10%): PKR {tax_correct:,}')
print(f'Net salary: PKR {net_salary:,.2f}')

# Lesson: when unsure, use parentheses. They make intent clear AND ensure correctness.
```

## 8. Walrus Operator `:=` (Python 3.8+)

The walrus operator `:=` is called that because it looks like a walrus lying on its side (`:=` — two eyes and two tusks). It **assigns a value AND returns it** in the same expression.

This is most useful inside `while` loops and `if` statements where you want to assign and check at the same time.

```python
# Without walrus operator — two lines
data = [1, 5, 3, 9, 2, 7]
total = sum(data)
if total > 20:
    print(f'Total {total} is greater than 20')

# With walrus operator — assign and check in one line
if (total := sum(data)) > 20:
    print(f'Total {total} is greater than 20')

# total is available here after the if statement
print(f'Calculated total: {total}')
```

```python
# Classic walrus use case: processing chunks until empty
import io

# Simulating reading a file line by line
lines = ['Line 1: Hello\n', 'Line 2: Python\n', 'Line 3: Walrus\n', '']
idx = 0

# Without walrus:
line = lines[idx]
while line:
    print(line.strip())
    idx += 1
    line = lines[idx] if idx < len(lines) else ''

print('---')

# With walrus — much cleaner for actual file reading:
# while chunk := file.read(1024):  # keeps reading until empty string
#     process(chunk)
print('Walrus pattern is excellent for: file reading, user prompts, API pagination')
```

## 9. Common Mistakes

```python
# MISTAKE 1: Using = instead of == in comparisons
x = 10

# This is an ASSIGNMENT, not a comparison:
# if x = 5:   # SyntaxError in Python! (some languages allow this bug)

# Correct comparison:
if x == 10:
    print('x is 10')

# Python 3 prevents this mistake by making = inside if a SyntaxError
# (The walrus := is the intentional exception)
```

```python
# MISTAKE 2: Confusing 'is' with '=='
a = [1, 2, 3]
b = [1, 2, 3]

# Beginner might write this thinking they're checking equality:
print('Using is:', a is b)    # False — they're different objects!
print('Using ==:', a == b)    # True  — same values

# The ONLY safe use of 'is' with literals is for None, True, False:
value = None
print(value is None)    # Correct pattern
```

```python
# MISTAKE 3: Integer division vs float division surprise

# In Python 3, / always gives float:
print(5 / 2)    # 2.5  (not 2!)
print(type(5/2))  # float

# For integer result, you must use //:
print(5 // 2)   # 2
print(type(5//2))  # int

# Floor division floors towards negative infinity (not towards zero!)
print(7 // 2)    # 3    (expected)
print(-7 // 2)   # -4   (surprise! -3.5 floors DOWN to -4, not towards 0)
```

## 11. Solutions

```python
# SOLUTION 1 - try it yourself first!
ticket_price = 650
budget       = 4000
snack_price  = 300

tickets_bought = budget // ticket_price
leftover       = budget % ticket_price
can_afford_snack = leftover >= snack_price

print(f'Tickets: {tickets_bought}')
print(f'Leftover: PKR {leftover}')
print(f'Can afford snack: {can_afford_snack}')
```

```python
# SOLUTION 2 - try it yourself first!
password = 'Secure@1234'

long_enough     = len(password) >= 8
has_digit       = any(c in '0123456789' for c in password)
has_special     = any(c in '!@#$%' for c in password)

criteria_met = long_enough + has_digit + has_special  # booleans as ints

if criteria_met == 3:
    strength = 'Strong'
elif criteria_met == 2:
    strength = 'Medium'
else:
    strength = 'Weak'

print(f'Password: {password}')
print(f'Strength: {strength} ({criteria_met}/3 criteria met)')
print(f'  Long enough (>=8): {long_enough}')
print(f'  Has digit:         {has_digit}')
print(f'  Has special char:  {has_special}')
```

```python
# SOLUTION 3 - try it yourself first!
numbers = [88, 76, 92, 45, 67, 83, 91, 55, 78, 64]

total   = 0
count   = 0
maximum = numbers[0]   # start with first element
minimum = numbers[0]

for num in numbers:
    total += num
    count += 1
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

average = total / count

print(f'Numbers: {numbers}')
print(f'Count:   {count}')
print(f'Sum:     {total}')
print(f'Max:     {maximum}')
print(f'Min:     {minimum}')
print(f'Average: {average:.2f}')
```

## 12. Mini-Project: Grade Calculator

```python
# ============================================================
# MINI-PROJECT: Grade Calculator
# ============================================================
# Takes marks in 5 subjects, computes average,
# determines pass/fail and assigns a letter grade.

print('=== GRADE CALCULATOR ===')
print('Enter your marks out of 100 for each subject:\n')

# --- Input section ---
subjects = ['Mathematics', 'Physics', 'Chemistry', 'English', 'Urdu']
marks    = []

for subject in subjects:
    raw  = input(f'{subject}: ')
    mark = float(raw)
    marks.append(mark)

# --- Calculations using operators ---
total   = 0
highest = marks[0]
lowest  = marks[0]

for m in marks:
    total   += m                    # accumulate sum with +=
    highest  = m if m > highest else highest   # track max
    lowest   = m if m < lowest  else lowest    # track min

average  = total / len(subjects)
passed   = average >= 40            # pass threshold

# --- Letter grade using comparisons ---
if average >= 90:
    letter, description = 'A+', 'Outstanding'
elif average >= 80:
    letter, description = 'A', 'Excellent'
elif average >= 70:
    letter, description = 'B', 'Good'
elif average >= 60:
    letter, description = 'C', 'Average'
elif average >= 50:
    letter, description = 'D', 'Below Average'
elif average >= 40:
    letter, description = 'E', 'Pass'
else:
    letter, description = 'F', 'Fail'

# --- Report card ---
print()
print('=' * 42)
print(f'{"REPORT CARD":^42}')
print('=' * 42)
for subject, mark in zip(subjects, marks):
    bar = '#' * int(mark // 10)    # simple bar using //
    print(f'{subject:<12}: {mark:>5.1f}  [{bar:<10}]')
print('-' * 42)
print(f'Total     : {total:.1f} / {len(subjects)*100}')
print(f'Average   : {average:.1f}%')
print(f'Highest   : {highest:.1f}')
print(f'Lowest    : {lowest:.1f}')
print(f'Grade     : {letter}  ({description})')
print(f'Status    : {"PASS" if passed else "FAIL"}')
print('=' * 42)
```
