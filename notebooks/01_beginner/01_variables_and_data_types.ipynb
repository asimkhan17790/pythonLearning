## 1. What is a Variable?

Think of a variable like a **labelled sticky note** you stick onto a piece of data. The label is the variable name, the data is the value. Any time you want that data back, you just say the label.

In Python, creating a variable is as simple as writing:
```
label = value
```
The `=` sign means **assignment** — "make this label point to this value". It does NOT mean "equals" in the mathematical sense.

````python
# Create a variable called 'city' and assign the value 'Karachi' to it
city = 'Karachi'

# Now use the variable — Python looks up what 'city' points to
print(city)          # Output: Karachi

# You can reassign a variable at any time
city = 'Lahore'
print(city)          # Output: Lahore  (the old value is gone)
````

````python
# Variables can hold many kinds of values
player_name  = 'Babar Azam'   # text
test_average = 45.26          # decimal number
matches_played = 92           # whole number
is_captain   = True           # yes/no flag

print(player_name, 'has played', matches_played, 'tests with an average of', test_average)
````

## 2. Strings

A **string** is any sequence of characters (letters, digits, symbols, spaces) surrounded by quotes. Think of it as *text*.

Key rules:
- Use single quotes `'...'` or double quotes `"..."` — both work the same way
- The opening and closing quote must match
- You can include the other quote type *inside* your string without escaping
- Strings are **immutable** — once created, the characters cannot be changed in place

````python
# Single vs double quotes — both are strings
first_name = 'Aisha'
last_name  = "Siddiqui"

# Using the other quote type inside
message = "It's a beautiful day"   # single quote inside double-quoted string — fine!
quote   = 'He said "hello" loudly' # double quote inside single-quoted string — fine!

print(message)
print(quote)
````

````python
# String concatenation: joining strings with +
full_name = first_name + ' ' + last_name
print('Full name:', full_name)

# String repetition: repeat a string with *
divider = '-' * 30
print(divider)

# len() tells you how many characters are in a string
print('Length of full name:', len(full_name))
````

## 3. Integers

An **integer** (`int`) is a whole number — positive, negative, or zero. No decimal point.

Key rules:
- No limit on size in Python (unlike most other languages)
- You can use underscores `_` as a thousands separator for readability (e.g., `1_000_000`)
- Supports all standard arithmetic

````python
population_pakistan = 230_000_000   # underscore makes large numbers readable
temperature_arctic  = -40           # negative integer
floors_in_building  = 0             # zero is valid

print('Pakistan population:', population_pakistan)
print('Arctic temperature:', temperature_arctic, 'degrees Celsius')

# Basic arithmetic with integers
apples_in_box   = 24
boxes_delivered = 15
total_apples = apples_in_box * boxes_delivered
print('Total apples:', total_apples)
````

````python
# Integer division and modulo — very useful in real programs
total_seconds = 3725

hours   = total_seconds // 3600   # // is floor division (whole number result)
minutes = (total_seconds % 3600) // 60  # % is modulo (remainder)
seconds = total_seconds % 60

print(f'{total_seconds} seconds = {hours}h {minutes}m {seconds}s')
````

## 4. Floats

A **float** (floating-point number) is a number *with a decimal point*. The name comes from how computers store them — the decimal point can "float" to different positions.

Key rules:
- Any number with a `.` is a float: `3.14`, `0.5`, `100.0`
- Float arithmetic can have tiny rounding errors (this is normal in all computers)
- Dividing two integers with `/` always gives a float in Python 3

````python
pi            = 3.14159
gravity       = 9.81      # m/s²
exchange_rate = 278.50    # PKR per USD

# Regular division always returns a float
half = 10 / 2
print(half)         # 5.0  — notice the .0, it's a float!
print(type(half))   # <class 'float'>

# Real calculation
usd_amount = 150
pkr_amount = usd_amount * exchange_rate
print(f'${usd_amount} = PKR {pkr_amount:,.2f}')  # :,.2f formats nicely
````

````python
# The floating-point rounding quirk — don't be alarmed!
result = 0.1 + 0.2
print(result)           # You might expect 0.3, but get 0.30000000000000004

# This is a known computer science issue (binary representation of decimals)
# For financial calculations, use the 'round()' function or the 'decimal' module
print(round(result, 2)) # 0.3 — rounded to 2 decimal places
````

## 5. Booleans

A **boolean** (`bool`) has only two possible values: `True` or `False`. Named after mathematician George Boole.

Think of it as a light switch — it's either on or off. Booleans are the result of comparisons and control the flow of your programs.

Key rules:
- Must be capitalised: `True` and `False` (not `true` or `false`)
- Booleans are actually a subtype of integers in Python: `True == 1` and `False == 0`
- Comparisons always return booleans

````python
is_raining    = True
has_umbrella  = False
is_logged_in  = True

print(is_raining)              # True
print(type(is_raining))        # <class 'bool'>

# Booleans from comparisons
age = 20
can_vote = age >= 18
print('Can vote:', can_vote)   # True

score = 45
passed = score >= 50
print('Passed:', passed)        # False
````

````python
# Booleans are secretly integers — useful trick for counting
print(True + True)     # 2  (True counts as 1)
print(False + True)    # 1
print(True * 10)       # 10

# Practical example: count how many conditions are met
has_passport  = True
has_visa      = True
has_ticket    = False
ready_to_fly  = has_passport + has_visa + has_ticket
print(f'Ready for travel: {ready_to_fly}/3 requirements met')
````

## 6. None — The Absence of a Value

`None` is Python's way of saying **"nothing"**, **"empty"**, or **"not yet assigned"**. It's like a blank form field — the field exists, but has no value yet.

Key rules:
- `None` is its own type: `NoneType`
- Always written with capital N: `None`
- Functions that don't explicitly return something return `None` by default
- Check for None with `is None`, not `== None`

````python
# Representing 'not yet known'
middle_name    = None    # this person might not have a middle name
delivery_date  = None    # order placed but not shipped yet
search_result  = None    # nothing found yet

print(middle_name)            # None
print(type(middle_name))      # <class 'NoneType'>

# The correct way to check for None
if middle_name is None:
    print('No middle name provided')

# Later, the value can be filled in
middle_name = 'Fatima'
print('Middle name updated to:', middle_name)
````

## 7. Checking Types with `type()`

Python lets you ask "what type is this?" at any time using the built-in `type()` function. This is incredibly useful for debugging.

````python
# type() returns the data type of any value or variable
print(type('hello'))      # <class 'str'>
print(type(42))           # <class 'int'>
print(type(3.14))         # <class 'float'>
print(type(True))         # <class 'bool'>
print(type(None))         # <class 'NoneType'>

# A common gotcha: '42' (string) vs 42 (int)
user_input = '42'         # input() always returns a string
actual_number = 42
print(type(user_input))   # <class 'str'>  -- cannot do math with this!
print(type(actual_number))# <class 'int'>  -- can do math with this
````

````python
# isinstance() — a better check for 'is this a certain type?'
score = 95.5
print(isinstance(score, float))    # True
print(isinstance(score, int))      # False
print(isinstance(score, (int, float)))  # True — check against multiple types
````

## 8. Multiple Assignment and Variable Swap

Python has elegant shortcuts for assigning multiple variables at once and for swapping their values — no temporary variable needed!

````python
# Assign the same value to multiple variables in one line
x = y = z = 0
print(x, y, z)   # 0 0 0

# Assign different values to multiple variables in one line
name, age, city = 'Sara', 25, 'Islamabad'
print(name)   # Sara
print(age)    # 25
print(city)   # Islamabad

# Unpack a list or tuple into variables
coordinates = (33.72, 73.04)   # latitude and longitude of Islamabad
lat, lon = coordinates
print(f'Latitude: {lat}, Longitude: {lon}')
````

````python
# Swapping two variables — the Pythonic way
player1_score = 150
player2_score = 230

print(f'Before swap: Player1={player1_score}, Player2={player2_score}')

# In most languages you need a temp variable. Python doesn't!
player1_score, player2_score = player2_score, player1_score

print(f'After swap:  Player1={player1_score}, Player2={player2_score}')

# Why this works: Python evaluates the RIGHT side fully first,
# creating a tuple (230, 150), then unpacks it into the left side.
````

## 9. Type Conversion

Python won't automatically mix types for you. If you want to add a number to a string, you must **convert** one of them first. This is called **explicit type conversion** (or casting).

| Function | Converts to | Example |
|----------|-------------|--------|
| `int(x)` | Integer | `int('42')` → `42` |
| `float(x)` | Float | `float('3.14')` → `3.14` |
| `str(x)` | String | `str(100)` → `'100'` |
| `bool(x)` | Boolean | `bool(0)` → `False` |

````python
# str -> int: converting text that looks like a number
age_as_text   = '25'         # came from user input or a file
age_as_number = int(age_as_text)
print(age_as_number + 5)     # 30  — now we can do math

# int -> str: converting number to text for display
score  = 98
report = 'Your score is: ' + str(score)   # must convert, can't + a string and int
print(report)

# float -> int: drops the decimal part (does NOT round)
price   = 99.99
rounded = int(price)
print(rounded)    # 99  — truncated, not rounded!
````

````python
# bool() — what values become True vs False?
# Falsy values: 0, 0.0, '', None, empty containers
# Everything else is truthy

print(bool(0))       # False  — zero is falsy
print(bool(1))       # True
print(bool(-99))     # True   — any non-zero number is truthy
print(bool(''))      # False  — empty string is falsy
print(bool('hi'))    # True   — non-empty string is truthy
print(bool(None))    # False
print(bool([]))      # False  — empty list is falsy
print(bool([1,2]))   # True
````

````python
# Conversion failures — not everything can be converted
try:
    bad_conversion = int('hello')   # 'hello' is not a number!
except ValueError as e:
    print('Error:', e)   # invalid literal for int() with base 10: 'hello'

try:
    also_bad = int('3.14')  # int() can't parse a float string directly
except ValueError as e:
    print('Error:', e)

# Fix: convert to float first, then to int
fixed = int(float('3.14'))
print(fixed)   # 3
````

## 10. Mutable vs Immutable — A Mental Model

This is one of the most important concepts in Python.

- **Immutable** objects *cannot be changed* after they are created. Strings, integers, floats, booleans, and None are all immutable. When you "change" a string variable, Python creates a new string object — it doesn't modify the old one.
- **Mutable** objects *can be changed* in place. Lists, dictionaries, and sets are mutable (covered in a later notebook).

For now, the key insight: **with immutable types, reassigning a variable doesn't affect other variables that pointed to the same value.**

````python
# Immutable in action: integers
a = 10
b = a        # b points to the same value as a
a = 20       # reassigning a creates a NEW integer object

print('a =', a)   # 20
print('b =', b)   # 10  — b is unaffected! It still points to the old 10.

# id() shows the memory address — different objects have different ids
x = 'hello'
y = x
print(id(x) == id(y))   # True — they point to the SAME object
x = x + ' world'        # creates a NEW string, x now points elsewhere
print(id(x) == id(y))   # False — x moved, y stayed
print(y)                 # 'hello' — y is unchanged
````

## 11. Naming Rules — Valid and Invalid Names

Variable names have strict rules and strong conventions in Python.

**Rules (enforced by Python — breaking these causes errors):**
- Can only contain letters, digits, and underscores (`_`)
- Cannot start with a digit
- Cannot be a Python keyword (`if`, `for`, `class`, `return`, etc.)
- Case-sensitive: `Name` and `name` are different variables

**Conventions (community standards — breaking these causes confusion):**
- Use `snake_case` for regular variables and functions: `player_score`, `first_name`
- Use `UPPER_CASE` for constants: `MAX_SPEED = 300`
- Use `CamelCase` for class names: `PlayerCharacter`
- Avoid single-letter names except for simple counters (`i`, `x`, `n`)

````python
# VALID variable names
user_name     = 'Ali'
age2          = 30         # digit is fine as long as it's not first
_private      = 'internal' # leading underscore signals 'private by convention'
MAX_RETRIES   = 3          # constant (by convention, not enforced)
camelCase     = True       # works but not Pythonic for variables

print(user_name, age2, MAX_RETRIES)
````

````python
# INVALID variable names — these would cause SyntaxError
# Uncomment one at a time to see the error:

# 2fast = True          # starts with a digit
# my-name = 'Ali'       # hyphens not allowed (Python thinks it's subtraction)
# for = 10              # 'for' is a reserved keyword
# user name = 'Bob'     # spaces not allowed

# Case sensitivity demo
Name = 'Alice'
name = 'Bob'
NAME = 'CHARLIE'
print(Name, name, NAME)   # These are three DIFFERENT variables!
````

## 12. Common Mistakes

These are the errors beginners make most often with variables and types. Study them carefully!

````python
# MISTAKE 1: Mixing strings and numbers without converting
# This is the #1 beginner error.

user_age = '22'   # imagine this came from input()

try:
    result = user_age + 5   # TypeError! Can't add str and int
except TypeError as e:
    print('Error:', e)

# FIX: convert before doing arithmetic
result = int(user_age) + 5
print('Next birthday age:', result)  # 27
````

````python
# MISTAKE 2: Using = when you mean ==
# = assigns a value. == compares two values.

x = 10

# This ASSIGNS 5 to x (doesn't compare!):
# x = 5   <-- this would change x

# This COMPARES x to 5 (returns True/False):
print(x == 10)   # True
print(x == 5)    # False

# In an if statement, use == (not =)
if x == 10:
    print('x is ten')
````

````python
# MISTAKE 3: Thinking float division always gives a whole number

total = 10
parts = 3
share = total / parts
print(share)           # 3.3333...  — not 3!
print(type(share))     # float, even though inputs were ints

# If you want integer division: use //
whole_share = total // parts
remainder   = total % parts
print(f'{total} divided into {parts} = {whole_share} each, {remainder} leftover')
````

## 15. Mini-Project: Personal Info Card

Bring it all together! We'll store information about a person using five different variable types and print a nicely formatted summary card.

````python
# ============================================================
# MINI-PROJECT: Personal Info Card
# ============================================================

# --- Store the information ---
name        = 'Zainab Malik'   # str  — full name
age         = 22               # int  — age in years
height      = 1.65             # float — height in metres
is_student  = True             # bool  — currently enrolled?
gpa         = 3.78             # float — grade point average (0.0 – 4.0)

# --- Derived calculations ---
height_cm   = height * 100                  # convert metres to centimetres
gpa_percent = (gpa / 4.0) * 100            # convert GPA to percentage
status      = 'Student' if is_student else 'Graduate'

# --- Print the card ---
border = '=' * 40
print(border)
print('       PERSONAL INFORMATION CARD')
print(border)
print(f'  Name    : {name}')
print(f'  Age     : {age} years old')
print(f'  Height  : {height} m  ({height_cm:.0f} cm)')
print(f'  Status  : {status}')
print(f'  GPA     : {gpa:.2f} / 4.00  ({gpa_percent:.1f}%)')
print(border)

# --- Type summary ---
print('\nData type summary:')
for var_name, value in [('name', name), ('age', age), ('height', height),
                         ('is_student', is_student), ('gpa', gpa)]:
    print(f'  {var_name:<12} = {repr(value):<15}  type: {type(value).__name__}')
````
