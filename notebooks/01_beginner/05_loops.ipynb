# Loops in Python

Loops let you **repeat a block of code** multiple times without writing it over and over.

Python has two kinds of loops:
- `for` loop — repeat a fixed number of times (iterate over a sequence)
- `while` loop — repeat as long as a condition is true

By the end of this notebook you will be able to:
- Use `for` and `while` loops confidently
- Control loops with `break`, `continue`, and `pass`
- Use `range()`, `enumerate()`, and `zip()`
- Write nested loops
- Avoid the most common loop mistakes

## 1. The `for` Loop

A `for` loop iterates over a **sequence** (list, string, range, etc.) and runs the body once per item.

```python
# Basic for loop over a list
fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    print(fruit)
```

```python
# Loop over a string (each character)
for char in 'Hello':
    print(char)
```

## 2. `range()` — Generate a Sequence of Numbers

`range(start, stop, step)` generates numbers from `start` up to (but not including) `stop`.

| Call | Result |
|------|--------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(2, 6)` | 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(5, 0, -1)` | 5, 4, 3, 2, 1 |

```python
# Count from 0 to 4
for i in range(5):
    print(i)

print('---')

# Count from 1 to 5
for i in range(1, 6):
    print(i)

print('---')

# Count down
for i in range(5, 0, -1):
    print(i)
```

## 3. The `while` Loop

A `while` loop keeps running **as long as its condition is True**. Use it when you don't know in advance how many iterations you need.

```python
# Count up to 5 using while
count = 1

while count <= 5:
    print(f'Count: {count}')
    count += 1  # IMPORTANT: always update the variable or you get an infinite loop!

print('Done!')
```

```python
# Real-world example: keep asking for input until valid
# (Simulated here — in a real program you'd use input())
responses = ['no', 'no', 'yes']  # simulate user responses
index = 0

answer = responses[index]
while answer != 'yes':
    print(f'User said: {answer}. Asking again...')
    index += 1
    answer = responses[index]

print(f'User said: {answer}. Moving on!')
```

## 4. `break` — Exit a Loop Early

`break` immediately exits the loop, regardless of the condition.

```python
# Stop as soon as we find the number 3
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num == 3:
        print('Found 3! Stopping.')
        break
    print(num)

# Output: 1, 2, Found 3! Stopping.
```

```python
# break in a while loop
secret = 42
guesses = [10, 25, 42, 99]  # simulate guesses

for guess in guesses:
    if guess == secret:
        print(f'Correct! The answer was {secret}')
        break
    else:
        print(f'{guess} is wrong, try again')
```

## 5. `continue` — Skip to the Next Iteration

`continue` skips the rest of the current iteration and moves to the next one.

```python
# Print only even numbers
for i in range(1, 11):
    if i % 2 != 0:  # if odd...
        continue    # ...skip it
    print(i)
```

```python
# Skip empty strings
names = ['Alice', '', 'Bob', '', 'Charlie']

for name in names:
    if name == '':
        continue
    print(f'Hello, {name}!')
```

## 6. `pass` — Do Nothing (Placeholder)

`pass` is a no-op. It's used as a placeholder when syntax requires a block but you have nothing to write yet.

```python
# pass as placeholder
for i in range(5):
    pass  # TODO: add logic later

print('Loop finished (pass did nothing)')
```

## 7. `enumerate()` — Loop with Index

When you need both the **index** and the **value** while looping, use `enumerate()`.

```python
fruits = ['apple', 'banana', 'cherry']

# Without enumerate (clunky)
for i in range(len(fruits)):
    print(f'{i}: {fruits[i]}')

print('---')

# With enumerate (clean)
for index, fruit in enumerate(fruits):
    print(f'{index}: {fruit}')

print('---')

# Start index from 1
for index, fruit in enumerate(fruits, start=1):
    print(f'{index}. {fruit}')
```

## 8. `zip()` — Loop Over Two Lists Together

`zip()` pairs up elements from two (or more) sequences.

```python
names = ['Alice', 'Bob', 'Charlie']
scores = [95, 82, 78]

for name, score in zip(names, scores):
    print(f'{name} scored {score}')

print('---')

# zip stops at the shortest list
a = [1, 2, 3, 4, 5]
b = ['x', 'y', 'z']  # shorter

for num, letter in zip(a, b):
    print(num, letter)  # only 3 pairs
```

## 9. Iterating Over Dictionaries

Dictionaries have three iteration methods: `.keys()`, `.values()`, `.items()`.

```python
student = {'name': 'Alice', 'age': 20, 'grade': 'A'}

# Iterate over keys (default)
print('Keys:')
for key in student:
    print(f'  {key}')

# Iterate over values
print('Values:')
for value in student.values():
    print(f'  {value}')

# Iterate over key-value pairs
print('Items:')
for key, value in student.items():
    print(f'  {key} = {value}')
```

## 10. Nested Loops

A loop inside a loop. The inner loop runs **completely** for each iteration of the outer loop.

```python
# Multiplication table (3x3)
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i} x {j} = {i*j}', end='   ')
    print()  # newline after each row
```

```python
# Nested loop with a list of lists
classroom = [
    ['Alice', 'Bob'],
    ['Charlie', 'Diana'],
    ['Eve', 'Frank']
]

for row_num, row in enumerate(classroom, 1):
    for student in row:
        print(f'Row {row_num}: {student}')
```

## 11. `for`...`else` and `while`...`else`

Python loops have an optional `else` clause that runs **only if the loop completed normally** (not exited via `break`).

```python
# Search for a number — use else to know if it was found
def find_number(numbers, target):
    for num in numbers:
        if num == target:
            print(f'Found {target}!')
            break
    else:
        # This runs only if break was never hit
        print(f'{target} not found in list')

find_number([1, 2, 3, 4, 5], 3)   # Found!
find_number([1, 2, 3, 4, 5], 99)  # Not found
```

## Common Mistakes

### Mistake 1: Infinite Loop — Forgetting to Update the Variable

```python
# BAD — this runs forever! (don't actually run this)
# count = 0
# while count < 5:
#     print(count)
#     # forgot: count += 1

# GOOD
count = 0
while count < 5:
    print(count)
    count += 1  # must update!
print('Done')
```

### Mistake 2: Modifying a List While Iterating Over It

```python
# BAD — unpredictable behavior
numbers = [1, 2, 3, 4, 5]
# for num in numbers:
#     if num == 3:
#         numbers.remove(num)  # modifying while iterating!

# GOOD — iterate over a copy
numbers = [1, 2, 3, 4, 5]
for num in numbers[:]:  # numbers[:] creates a copy
    if num == 3:
        numbers.remove(num)

print(numbers)  # [1, 2, 4, 5]
```

### Mistake 3: Using `range(len(list))` When `enumerate()` Is Cleaner

```python
items = ['a', 'b', 'c']

# BAD (works but ugly)
for i in range(len(items)):
    print(i, items[i])

print('---')

# GOOD
for i, item in enumerate(items):
    print(i, item)
```

### Exercise 3 (Hard): Find Prime Numbers
Print all prime numbers between 1 and 50 using nested loops.

```python
# Your solution here
```

```python
# Solution
primes = []
for num in range(2, 51):
    is_prime = True
    for divisor in range(2, int(num**0.5) + 1):
        if num % divisor == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print('Primes between 1 and 50:', primes)
```

## Mini-Project: Number Guessing Game

Build a simple number guessing game. The program picks a secret number and tells the player if their guess is too high, too low, or correct. Limit to 5 attempts.

```python
import random

def play_guessing_game(secret=None):
    """Number guessing game with limited attempts."""
    if secret is None:
        secret = random.randint(1, 100)
    
    max_attempts = 5
    print(f'Guess the number between 1 and 100! You have {max_attempts} attempts.')
    
    for attempt in range(1, max_attempts + 1):
        # In a real game: guess = int(input(f'Attempt {attempt}: '))
        # For demo, we simulate guesses
        import random as r
        guess = r.randint(1, 100)
        print(f'Attempt {attempt}: Guessing {guess}...')
        
        if guess < secret:
            print('  Too low!')
        elif guess > secret:
            print('  Too high!')
        else:
            print(f'  Correct! You got it in {attempt} attempts!')
            break
    else:
        print(f'Out of attempts! The number was {secret}.')

# Play with a known secret so output is deterministic for demo
play_guessing_game(secret=42)
```
