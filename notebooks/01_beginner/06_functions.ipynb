# Functions in Python

A **function** is a reusable block of code that performs a specific task.

Instead of writing the same code repeatedly, you define it once and **call** it whenever needed.

By the end of this notebook you will be able to:
- Define and call functions
- Use parameters, default values, `*args`, and `**kwargs`
- Return values from functions
- Write lambda (anonymous) functions
- Understand closures and first-class functions
- Write clean, well-documented functions

## 1. Defining and Calling a Function

Use the `def` keyword, followed by the function name and parentheses `()`.

````python
# Define a function
def greet():
    print('Hello, World!')

# Call the function
greet()
greet()  # can call it many times
````

````python
# Function with a parameter
def greet_person(name):
    print(f'Hello, {name}!')

greet_person('Alice')
greet_person('Bob')
````

## 2. Parameters vs Arguments

- **Parameter**: the variable name in the function *definition* → `name` in `def greet(name)`
- **Argument**: the actual value you *pass* when calling → `'Alice'` in `greet('Alice')`

### Positional Arguments
Matched by position — first argument → first parameter.

````python
def describe_pet(animal, name):
    print(f'I have a {animal} named {name}.')

describe_pet('dog', 'Rex')    # positional: animal='dog', name='Rex'
describe_pet('cat', 'Whiskers')

# Keyword arguments — order doesn't matter
describe_pet(name='Goldie', animal='fish')
````

## 3. Default Parameter Values

Give a parameter a **default value** so callers can omit it.

````python
def power(base, exponent=2):  # exponent defaults to 2
    return base ** exponent

print(power(3))     # 3^2 = 9 (uses default)
print(power(3, 3))  # 3^3 = 27
print(power(2, 10)) # 2^10 = 1024
````

````python
# Real-world: greeting with optional title
def greet(name, title='Mr./Ms.'):
    return f'Hello, {title} {name}!'

print(greet('Smith'))
print(greet('Johnson', 'Dr.'))
````

## 4. Return Values

Use `return` to send a value back to the caller.
- Without `return`, a function returns `None`
- You can return **multiple values** (returned as a tuple)

````python
def add(a, b):
    return a + b

result = add(3, 4)
print(result)  # 7

# Use the result directly
print(add(10, 20) * 2)  # 60
````

````python
# Returning multiple values
def min_max(numbers):
    return min(numbers), max(numbers)  # returns a tuple

low, high = min_max([3, 1, 4, 1, 5, 9, 2, 6])
print(f'Min: {low}, Max: {high}')
````

## 5. `*args` — Variable Number of Positional Arguments

`*args` collects any extra positional arguments into a **tuple**.

````python
def add_all(*args):
    print(f'args = {args}')  # it's a tuple
    return sum(args)

print(add_all(1, 2))          # 3
print(add_all(1, 2, 3, 4))    # 10
print(add_all(10, 20, 30, 40, 50))  # 150
````

````python
# Mixing regular params and *args
def introduce(greeting, *names):
    for name in names:
        print(f'{greeting}, {name}!')

introduce('Hello', 'Alice', 'Bob', 'Charlie')
````

## 6. `**kwargs` — Variable Number of Keyword Arguments

`**kwargs` collects extra keyword arguments into a **dictionary**.

````python
def print_info(**kwargs):
    print(f'kwargs = {kwargs}')  # it's a dict
    for key, value in kwargs.items():
        print(f'  {key}: {value}')

print_info(name='Alice', age=30, city='Paris')
````

````python
# Full combination: positional, *args, keyword, **kwargs
def full_example(required, *args, default='default', **kwargs):
    print(f'required: {required}')
    print(f'args: {args}')
    print(f'default: {default}')
    print(f'kwargs: {kwargs}')

full_example('must', 1, 2, 3, default='custom', extra='bonus', flag=True)
````

## 7. Docstrings

A **docstring** is a string at the start of a function that documents what it does.  
It's accessible via `help()` and documentation tools.

````python
def calculate_bmi(weight_kg, height_m):
    """
    Calculate Body Mass Index (BMI).

    Args:
        weight_kg (float): Weight in kilograms.
        height_m (float): Height in metres.

    Returns:
        float: The BMI value.
    """
    return weight_kg / (height_m ** 2)

bmi = calculate_bmi(70, 1.75)
print(f'BMI: {bmi:.1f}')

# View the docstring
help(calculate_bmi)
````

## 8. Lambda Functions

A **lambda** is a small, anonymous (nameless) function defined in one line.

```python
lambda parameters: expression
```

Useful for short, throwaway functions — especially as arguments to `sorted()`, `map()`, `filter()`.

````python
# Regular function vs lambda
def square(x):
    return x ** 2

square_lambda = lambda x: x ** 2

print(square(5))        # 25
print(square_lambda(5)) # 25

# Lambda with two args
add = lambda a, b: a + b
print(add(3, 4))  # 7

# Lambda as argument to sorted()
people = [('Alice', 30), ('Bob', 25), ('Charlie', 35)]
sorted_by_age = sorted(people, key=lambda person: person[1])
print(sorted_by_age)
````

## 9. Functions are First-Class Objects

In Python, functions are **objects** — you can:
- Assign them to variables
- Pass them as arguments
- Return them from other functions

````python
def shout(text):
    return text.upper() + '!'

def whisper(text):
    return text.lower() + '...'

def say_it(func, message):
    """Takes a function and applies it to message."""
    result = func(message)
    print(result)

say_it(shout, 'hello')    # HELLO!
say_it(whisper, 'HELLO')  # hello...

# Store function in a variable
my_func = shout
print(my_func('world'))   # WORLD!
````

## 10. Closures

A **closure** is a function that **remembers** the variables from its enclosing scope even after that scope has finished.

````python
def make_multiplier(n):
    """Returns a function that multiplies by n."""
    def multiply(x):
        return x * n  # 'n' is captured from the outer scope
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))  # 10
print(triple(5))  # 15
print(double(triple(4)))  # double(12) = 24
````

## Common Mistakes

### Mistake 1: Mutable Default Arguments

````python
# BAD — the list is shared across all calls!
def append_to_bad(item, my_list=[]):
    my_list.append(item)
    return my_list

print(append_to_bad(1))  # [1]
print(append_to_bad(2))  # [1, 2] — NOT [2]! Bug!

# GOOD — use None as default, create list inside
def append_to_good(item, my_list=None):
    if my_list is None:
        my_list = []
    my_list.append(item)
    return my_list

print(append_to_good(1))  # [1]
print(append_to_good(2))  # [2] — correct!
````

### Mistake 2: Not Using the Return Value

````python
def add(a, b):
    return a + b

# BAD — result is thrown away
add(3, 4)  # result ignored

# GOOD — capture it
result = add(3, 4)
print(result)

# Also good — use it immediately
print(add(3, 4))
````

### Mistake 3: Overwriting Built-in Names

````python
# BAD — shadows the built-in 'print'
# def print(text):
#     pass  # now built-in print is gone!

# BAD — shadows built-in 'list'
# list = [1, 2, 3]  # now you can't use list() anymore!

# GOOD — use descriptive names
def display_message(text):
    print(text)

my_list = [1, 2, 3]
print('Names are fine:', my_list)
````

### Exercise 2 (Medium): Flexible Statistics
Write a function `stats(*numbers)` that accepts any number of values and returns a dict with `min`, `max`, `sum`, and `average`.

````python
# Your solution here
````

````python
# Solution
def stats(*numbers):
    if not numbers:
        return None
    return {
        'min': min(numbers),
        'max': max(numbers),
        'sum': sum(numbers),
        'average': sum(numbers) / len(numbers)
    }

result = stats(5, 3, 8, 1, 9, 2)
for key, value in result.items():
    print(f'{key}: {value}')
````

### Exercise 3 (Hard): Function Factory
Write a function `make_greeting(language)` that returns a greeting function based on the language ('english', 'spanish', 'french').  
The returned function should accept a name and print the appropriate greeting.

````python
# Your solution here
````

````python
# Solution
def make_greeting(language):
    greetings = {
        'english': 'Hello',
        'spanish': 'Hola',
        'french': 'Bonjour'
    }
    phrase = greetings.get(language, 'Hello')  # default to English
    
    def greet(name):
        print(f'{phrase}, {name}!')
    
    return greet

say_hello = make_greeting('english')
say_hola = make_greeting('spanish')
say_bonjour = make_greeting('french')

say_hello('Alice')
say_hola('Carlos')
say_bonjour('Marie')
````

## Mini-Project: Calculator

Build a calculator that supports add, subtract, multiply, divide — using functions and a dispatch dictionary.

````python
def add(a, b):      return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b):
    if b == 0:
        return 'Error: division by zero'
    return a / b

# Map operator symbols to functions
operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide
}

def calculate(a, operator, b):
    """Perform a calculation given two numbers and an operator string."""
    if operator not in operations:
        return f'Unknown operator: {operator}'
    func = operations[operator]
    result = func(a, b)
    print(f'{a} {operator} {b} = {result}')
    return result

# Test the calculator
calculate(10, '+', 5)
calculate(10, '-', 3)
calculate(4, '*', 7)
calculate(15, '/', 3)
calculate(10, '/', 0)  # division by zero
calculate(5, '^', 2)   # unknown operator
````
