## 1. `print()` Basics — sep and end

`print()` is your main tool for displaying output. By default:
- Items separated by a comma print with a **space** between them
- Each `print()` call ends with a **newline** (moves to the next line)

You can customise both behaviours:
- `sep` — what to put **between** items (default: `' '`)
- `end` — what to put at the **end** (default: `'\n'`)

````python
# Default behaviour
print('Hello', 'World', 'Python')   # Hello World Python  (spaces between)

# Custom separator
print('2024', '03', '15', sep='-')  # 2024-03-15  (date format)
print('apple', 'banana', 'cherry', sep=' | ')  # apple | banana | cherry
print('H', 'e', 'l', 'l', 'o', sep='')         # Hello  (no separator)
````

````python
# Custom end — useful for printing on the same line
print('Loading', end='')           # no newline
print('...', end='')               # still same line
print(' Done!')                    # now the newline happens

# Progress bar simulation
import time
print('Progress: ', end='')
for i in range(1, 6):
    print(f'[{i*20}%]', end=' ')   # all on one line
print()  # final newline

# Printing nothing — blank line
print('Line 1')
print()             # blank line
print('Line 3')
````

````python
# Printing multiple items vs using sep
name  = 'Hamza'
score = 95
grade = 'A+'

# Three ways to print the same information:
print('Student:', name, 'Score:', score, 'Grade:', grade)
print(f'Student: {name}  Score: {score}  Grade: {grade}')   # f-string (covered next)
print('Student:', name, 'Score:', score, 'Grade:', grade, sep=' | ')  # custom sep
````

## 2. f-strings — The Modern Way

**f-strings** (formatted string literals) are the best way to embed variables and expressions inside strings. Introduced in Python 3.6.

- Start the string with the letter `f` before the opening quote: `f'...'`
- Place any variable or expression inside `{}` curly braces
- Python evaluates what's inside `{}` and inserts the result

Think of `{}` as a **window** cut into your string where Python can look in and place a value.

````python
# Basic f-string: embed a variable
city      = 'Lahore'
temp_c    = 38
print(f'Current temperature in {city}: {temp_c} degrees Celsius')

# Embed an expression (not just a variable)
temp_f = (temp_c * 9/5) + 32    # formula inside the {}
print(f'That is {(temp_c * 9/5) + 32:.1f} degrees Fahrenheit')

# Embed function calls
name = 'ali raza'
print(f'Formatted name: {name.title()}')  # title() capitalises each word
````

````python
# You can put any Python expression inside {}
a, b = 7, 3
print(f'{a} + {b} = {a + b}')
print(f'{a} * {b} = {a * b}')
print(f'{a} ** {b} = {a ** b}')  # exponent

# Conditional expressions inside f-strings
marks = 72
print(f'Result: {"Pass" if marks >= 50 else "Fail"}')

# List indexing inside f-strings
fruits = ['mango', 'guava', 'lychee']
print(f'Favourite fruit: {fruits[0]}')
````

## 3. f-string Formatting — Numbers, Percentages, Padding, Alignment

The real power of f-strings comes from **format specifiers** — instructions after a `:` inside `{}` that control how the value is displayed.

```
f'{value:format_spec}'
         ↑
         colon starts the format specification
```

````python
# ---- Number formatting ----
pi       = 3.14159265
salary   = 85000.0
big_num  = 1234567.89

print(f'Pi to 2 decimal places:   {pi:.2f}')      # .2f = 2 decimal places, fixed
print(f'Pi to 4 decimal places:   {pi:.4f}')
print(f'Salary with commas:        {salary:,.2f}') # , adds thousands separator
print(f'Big number:                {big_num:,.0f}') # 0 decimal places
print(f'Scientific notation:       {big_num:.2e}')  # e notation
print(f'Integer only:              {pi:.0f}')       # rounds to nearest int
````

````python
# ---- Percentage formatting ----
pass_rate    = 0.876
tax_rate     = 0.175
discount     = 0.33333

print(f'Exam pass rate: {pass_rate:.1%}')   # .1% → multiplies by 100, adds %
print(f'Tax rate:       {tax_rate:.2%}')
print(f'Discount:       {discount:.0%}')    # rounded percentage
````

````python
# ---- Padding and alignment ----
# Format: f'{value:<width}' for left, f'{value:>width}' for right, f'{value:^width}' for centre

items = [('Laptop', 85000), ('Mouse', 1500), ('Keyboard', 3200), ('Monitor', 32000)]

print(f'{'Item':<15} {'Price (PKR)':>12}')
print('-' * 28)
for name, price in items:
    print(f'{name:<15} {price:>12,}')   # left-align name, right-align price with commas

print()
# Centre alignment with a fill character
print(f'{"REPORT TITLE":=^40}')   # fill with =, centre, width 40
````

````python
# ---- Zero padding for numbers ----
# Useful for things like roll numbers, IDs, timestamps
roll_number = 7
print(f'Roll No: {roll_number:03d}')    # 007  (pad with zeros to 3 digits)

for hour in range(0, 24, 6):
    print(f'Time: {hour:02d}:00')       # 00:00, 06:00, 12:00, 18:00
````

## 4. Old-Style Formatting — % and .format() (Historical Context)

Before f-strings existed, Python programmers used two other methods. You will see these in older code, tutorials, and libraries — so it's worth knowing what they look like.

````python
# ---- % formatting (oldest style, from Python 2 era) ----
# Uses placeholders like %s (string), %d (integer), %f (float)

name  = 'Nadia'
score = 88.5

print('Name: %s, Score: %.1f' % (name, score))   # % operator with a tuple

# Common % placeholders:
# %s  = string
# %d  = integer
# %f  = float  (%.2f = 2 decimal places)
# %05d = zero-padded integer to width 5
print('ID: %05d' % 42)   # ID: 00042
````

````python
# ---- .format() method (Python 3 style, before f-strings) ----
# Uses {} as placeholders, values passed as arguments to .format()

template = 'Hello, {}! You scored {} out of {}.'
print(template.format('Omar', 45, 50))

# Named placeholders — clearer for many variables
print('Dear {name}, your account balance is PKR {balance:,.2f}.'.format(
    name='Ms. Fatima',
    balance=24500.75
))

# Positional references (reuse a value)
print('{0} + {0} = {1}'.format(5, 10))   # 5 + 5 = 10
````

````python
# Quick comparison — all three produce the same output:
city  = 'Peshawar'
pop   = 2_200_000

print('%s has a population of %d' % (city, pop))               # % style
print('{} has a population of {}'.format(city, pop))           # .format() style
print(f'{city} has a population of {pop:,}')                   # f-string style (best)

# Recommendation: use f-strings for all NEW code you write.
````

## 6. `input()` Always Returns a String — Converting Input

This is a critical rule that trips up almost every beginner: **`input()` ALWAYS returns a string**, even if the user types a number.

If you need a number, you must **convert** it yourself.

````python
# Demonstrating the problem
raw = input('Enter your age: ')      # user types 25
print(type(raw))                     # <class 'str'> -- it's a string, not a number!
print(repr(raw))                     # '25'  -- notice the quotes

# This would be WRONG — string + string, not addition:
# print(raw + 10)   # TypeError!

# CORRECT: convert first
age = int(raw)
print(f'In 10 years you will be {age + 10}')
````

````python
# Common pattern: convert inline
height = float(input('Enter your height in metres (e.g. 1.75): '))
weight = float(input('Enter your weight in kg (e.g. 70): '))

bmi = weight / (height ** 2)
print(f'Your BMI is: {bmi:.1f}')

if bmi < 18.5:
    category = 'Underweight'
elif bmi < 25.0:
    category = 'Normal weight'
elif bmi < 30.0:
    category = 'Overweight'
else:
    category = 'Obese'

print(f'Category: {category}')
````

````python
# Safe input: handle non-numeric input gracefully
raw_score = input('Enter your exam score (0-100): ')

try:
    score = int(raw_score)
    print(f'Score accepted: {score}')
except ValueError:
    print(f'Error: "{raw_score}" is not a valid number. Please enter digits only.')
````

## 7. Multi-line Strings

Sometimes your text spans multiple lines. Python offers two ways to handle this:
1. **Triple quotes** (`'''...'''` or `"""..."""`) — the string spans multiple lines literally
2. **Escape characters** — use `\n` for a newline inside a regular string

````python
# Triple-quoted string — what you type is what you get
poem = """Roses are red,
Violets are blue,
Python is awesome,
And so are you!"""

print(poem)
print('---')
print('Number of lines:', poem.count('\n') + 1)
````

````python
# Escape characters in regular strings
# \n = newline, \t = tab, \\ = backslash, \' = single quote inside single-quoted string

address = 'House 42\nStreet 7, Gulberg\nLahore, Pakistan'
print(address)
print()

# Tab character for column alignment
print('Name\tAge\tCity')
print('Ahmed\t25\tKarachi')
print('Sara\t22\tLahore')

# Backslash in file paths (Windows style)
path = 'C:\\Users\\Ahmed\\Documents\\file.txt'
print(path)
````

````python
# Triple-quoted f-string — combine multiline and formatting
product  = 'Wireless Headphones'
price    = 4500
in_stock = True

description = f"""
Product: {product}
Price:   PKR {price:,}
Status:  {'In Stock' if in_stock else 'Out of Stock'}
"""
print(description)
````

## 8. Raw Strings

A **raw string** treats backslashes `\` as literal characters — they are NOT interpreted as escape sequences. Prefix with `r`.

This is especially useful for:
- Windows file paths: `r'C:\Users\name\file.txt'`
- Regular expressions (patterns for searching text)

````python
# Normal string: \n becomes a newline, \t becomes a tab
normal = 'C:\new_folder\table.csv'
print('Normal string:')
print(normal)   # The \n and \t are interpreted as newline and tab — broken path!

print()

# Raw string: backslashes are kept as-is
raw_path = r'C:\new_folder\table.csv'
print('Raw string:')
print(raw_path)   # C:\new_folder\table.csv  — correct!

print()
# Alternative: double every backslash
doubled = 'C:\\new_folder\\table.csv'
print('Doubled backslashes:')
print(doubled)   # Same result as raw string
````

````python
# Raw strings with regular expressions
import re

text = 'Call us at 0300-1234567 or 021-35678900'

# Without raw string, \d would need to be \\d
pattern = r'\d{3,4}-\d{7,8}'   # pattern to match phone numbers
phones = re.findall(pattern, text)
print('Phone numbers found:', phones)
````

## 9. Common Mistakes

````python
# MISTAKE 1: Forgetting that input() returns a string
# One of the most common beginner errors.

num1 = input('Enter first number: ')   # user types 5
num2 = input('Enter second number: ')  # user types 3

# BUG: this does string concatenation, not addition!
wrong_result = num1 + num2
print('Wrong result:', wrong_result)   # prints '53', not 8 !

# FIX: convert to int first
correct_result = int(num1) + int(num2)
print('Correct result:', correct_result)  # prints 8
````

````python
# MISTAKE 2: Forgetting the f before an f-string

name  = 'Kamran'
score = 95

# Without the f prefix — the {} are printed literally!
bad_output = 'Hello {name}, your score is {score}'
print(bad_output)   # Prints: Hello {name}, your score is {score}  — WRONG!

# With the f prefix — works correctly
good_output = f'Hello {name}, your score is {score}'
print(good_output)  # Prints: Hello Kamran, your score is 95
````

````python
# MISTAKE 3: Confusing print() with return
# print() just DISPLAYS a value — it doesn't save it for later use.

# This is NOT how you save a value:
# x = print(42)  -- this gives x the value None, not 42!

x = print(42)       # prints 42 to screen
print('x is:', x)   # x is: None  -- print() returns None!

# To save a value, just assign it directly:
x = 42
print('x is:', x)   # x is: 42
````

````python
# SOLUTION 2 - try it yourself first!
name     = input('Your name: ')
salary   = float(input('Monthly salary (PKR): '))
expenses = float(input('Monthly expenses (PKR): '))

savings        = salary - expenses
savings_pct    = savings / salary

print()
print(f'{'Monthly Financial Summary for ' + name:=^45}')
print(f'  Income   : PKR {salary:>12,.2f}')
print(f'  Expenses : PKR {expenses:>12,.2f}')
print(f'  Savings  : PKR {savings:>12,.2f}  ({savings_pct:.1%})')
print('=' * 45)
````

````python
# SOLUTION 3 - try it yourself first!
items = [
    ('Rice (5kg bag)', 2, 850.00),
    ('Cooking Oil (1L)', 3, 420.00),
    ('Sugar (1kg)', 4, 160.00),
    ('Milk (pack)', 6, 185.00),
    ('Bread', 2, 95.00),
]

print(f'{'Item':<22} {'Qty':>5} {'Unit Price':>12} {'Total':>10}')
print('-' * 52)

grand_total = 0
for name, qty, unit_price in items:
    total = qty * unit_price
    grand_total += total
    print(f'{name:<22} {qty:>5} {unit_price:>12.2f} {total:>10.2f}')

print('-' * 52)
print(f'{'GRAND TOTAL':<28} {grand_total:>10.2f} PKR')
````

## 12. Mini-Project: Mad Libs

Mad Libs is a word game where you fill in blanks with different types of words, then read the resulting (often silly) story. We'll prompt the user for 5 different inputs and weave them into a story using f-strings.

````python
# ============================================================
# MINI-PROJECT: Mad Libs Story Generator
# ============================================================

print('=== MAD LIBS: The Python Adventure ===')
print('Answer the prompts below to create your story!\n')

# --- Collect user inputs ---
hero_name     = input('Enter a hero name: ')
magical_item  = input('Enter a magical item (e.g. sword, staff, laptop): ')
villain_name  = input('Enter a villain name: ')
place         = input('Enter a place (e.g. mountain, office, marketplace): ')
superpower    = input('Enter a superpower (e.g. flying, coding, making tea): ')

# --- Derived details ---
story_length  = len(hero_name) * 7   # silly 'length' based on name
chapters      = max(3, len(magical_item))  # at least 3 chapters

# --- Print the story ---
print()
print('*' * 50)
print(f"{'THE TALE OF ' + hero_name.upper():^50}")
print('*' * 50)
print(f"""
Once upon a time, in the mystical land of {place.title()},
there lived a brave hero named {hero_name.title()}.

{hero_name.title()} possessed a legendary {magical_item}, which glowed
with an eerie light whenever danger was near.

One stormy evening, the fearsome villain {villain_name.title()} appeared
at the gates of {place.title()}, demanding tribute from all the citizens.

But {hero_name.title()} stood firm. With the power of {superpower},
and armed with the {magical_item}, our hero charged forward!

The battle lasted {story_length} minutes and spanned {chapters} chapters
of glorious combat. In the end, {hero_name.title()} prevailed,
and {villain_name.title()} fled, never to return.

And so, peace was restored to {place.title()}.
              --- THE END ---
""")
print('*' * 50)
print(f'Story stats: {len(hero_name + magical_item + villain_name + place + superpower)} characters of input')
````
