# Data Visualization with Matplotlib

**Matplotlib** is Python's most popular visualization library. It creates static, animated, and interactive plots.

Install: `pip install matplotlib`

The standard import:
```python

```

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f'Matplotlib version: {matplotlib.__version__}')

# Set default style
plt.style.use('seaborn-v0_8')
print('Matplotlib ready!')
````

## 1. Line Plot

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 4))
plt.plot(x, y_sin, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', color='red', linewidth=2, linestyle='--')

plt.title('Sine and Cosine Waves', fontsize=14)
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 2. Bar Chart

````python
import matplotlib.pyplot as plt

languages = ['Python', 'JavaScript', 'Java', 'C++', 'Go', 'Rust']
popularity = [30.5, 25.8, 18.2, 10.1, 8.7, 6.7]
colors = ['#3776AB', '#F7DF1E', '#ED8B00', '#00599C', '#00ADD8', '#CE412B']

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(languages, popularity, color=colors, edgecolor='white', linewidth=0.8)

# Add value labels on bars
for bar, val in zip(bars, popularity):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val}%', ha='center', va='bottom', fontweight='bold')

ax.set_title('Programming Language Popularity 2024', fontsize=14, pad=15)
ax.set_xlabel('Language')
ax.set_ylabel('Popularity (%)')
ax.set_ylim(0, 38)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
````

## 3. Histogram

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=1000)  # heights in cm

fig, ax = plt.subplots(figsize=(9, 5))
n, bins, patches = ax.hist(data, bins=30, color='steelblue', edgecolor='white', alpha=0.8)

# Add mean and std lines
ax.axvline(data.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {data.mean():.1f}cm')
ax.axvline(data.mean() + data.std(), color='orange', linestyle=':', label=f'+1 SD: {data.mean()+data.std():.1f}cm')
ax.axvline(data.mean() - data.std(), color='orange', linestyle=':')

ax.set_title('Distribution of Heights (N=1000)', fontsize=14)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Frequency')
ax.legend()
plt.tight_layout()
plt.show()
````

## 4. Scatter Plot

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 200
height = np.random.normal(170, 10, n)
weight = height * 0.45 - 5 + np.random.normal(0, 5, n)
age = np.random.randint(18, 65, n)

fig, ax = plt.subplots(figsize=(9, 6))
scatter = ax.scatter(height, weight, c=age, cmap='viridis', alpha=0.7, s=50)
plt.colorbar(scatter, ax=ax, label='Age (years)')

ax.set_title('Height vs Weight (colored by Age)', fontsize=14)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (kg)')
plt.tight_layout()
plt.show()
````

## 5. Subplots — Multiple Charts

````python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Multiple Plot Types', fontsize=16, fontweight='bold')

# Top-left: Line plot
x = np.linspace(0, 4*np.pi, 200)
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Line: sin(x)')

# Top-right: Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 67, 34]
axes[0, 1].bar(categories, values, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[0, 1].set_title('Bar Chart')

# Bottom-left: Scatter
np.random.seed(0)
x2 = np.random.randn(100)
y2 = x2 * 2 + np.random.randn(100)
axes[1, 0].scatter(x2, y2, alpha=0.5, color='purple')
axes[1, 0].set_title('Scatter Plot')

# Bottom-right: Pie chart
sizes = [30, 25, 20, 15, 10]
labels = ['Python', 'JS', 'Java', 'C++', 'Other']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
axes[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
````

## 6. Customization

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 5))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12, 15, 18, 22, 28, 35, 40, 38, 30, 25, 20, 15]

ax.plot(months, sales, 'o-', color='#2196F3', linewidth=2.5,
        markersize=8, markerfacecolor='white', markeredgewidth=2)
ax.fill_between(months, sales, alpha=0.1, color='#2196F3')

# Annotate peak
peak_idx = sales.index(max(sales))
ax.annotate(f'Peak: {max(sales)}k',
            xy=(months[peak_idx], max(sales)),
            xytext=(peak_idx - 2, max(sales) + 3),
            arrowprops=dict(arrowstyle='->', color='red'),
            color='red', fontweight='bold')

ax.set_title('Monthly Sales 2024', fontsize=15, fontweight='bold')
ax.set_ylabel('Sales (£k)')
ax.set_ylim(0, 50)
ax.grid(axis='y', alpha=0.4)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
````

## Mini-Project: Sales Dashboard

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
products = {
    'Laptops':  [45, 52, 48, 60, 55, 72],
    'Phones':   [120, 115, 130, 140, 125, 160],
    'Tablets':  [30, 28, 35, 32, 40, 45],
}
regions = ['North', 'South', 'East', 'West']
region_sales = [35, 28, 22, 15]

fig = plt.figure(figsize=(14, 10))
fig.suptitle('Sales Dashboard — Q1/Q2 2024', fontsize=16, fontweight='bold')

# Top: Multi-line chart
ax1 = fig.add_subplot(2, 2, (1, 2))
colors = ['#2196F3', '#4CAF50', '#FF9800']
for (product, sales), color in zip(products.items(), colors):
    ax1.plot(months, sales, 'o-', label=product, color=color, linewidth=2)
ax1.set_title('Monthly Sales by Product')
ax1.set_ylabel('Units Sold')
ax1.legend()
ax1.grid(alpha=0.3)

# Bottom-left: Stacked bar
ax2 = fig.add_subplot(2, 2, 3)
x = np.arange(len(months))
width = 0.6
bottom = np.zeros(len(months))
for (product, sales), color in zip(products.items(), colors):
    ax2.bar(x, sales, width, bottom=bottom, label=product, color=color, alpha=0.8)
    bottom += np.array(sales)
ax2.set_title('Stacked Units by Product')
ax2.set_xticks(x)
ax2.set_xticklabels(months)
ax2.set_ylabel('Units')
ax2.legend(loc='upper left', fontsize=8)

# Bottom-right: Pie by region
ax3 = fig.add_subplot(2, 2, 4)
ax3.pie(region_sales, labels=regions, autopct='%1.1f%%', startangle=90,
        colors=['#2196F3', '#4CAF50', '#FF9800', '#E91E63'])
ax3.set_title('Revenue by Region (%)')

plt.tight_layout()
plt.show()
print('Dashboard rendered!')
````
