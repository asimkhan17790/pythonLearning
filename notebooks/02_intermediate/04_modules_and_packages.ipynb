# Modules and Packages

A **module** is a Python file (`.py`) containing reusable code — functions, classes, and variables.

A **package** is a directory containing multiple modules plus an `__init__.py` file.

Modules let you:
- Organize code into logical units
- Reuse code across multiple projects
- Access Python's huge standard library and third-party ecosystem

By the end of this notebook you will be able to:
- Import modules in multiple ways
- Use the most important standard library modules
- Understand the `__name__ == '__main__'` guard
- Structure your own packages

## 1. Import Basics

There are four common ways to import.

```python
# 1. Import the whole module
import math
print(math.pi)        # 3.141592653589793
print(math.sqrt(16))  # 4.0
print(math.floor(3.7)) # 3

# 2. Import specific names
from math import pi, sqrt, factorial
print(pi)
print(sqrt(25))
print(factorial(5))  # 120

# 3. Import with alias
import math as m
print(m.ceil(3.2))   # 4

# 4. Import everything (not recommended in production)
# from math import *  # pollutes namespace
```

## 2. The `math` Module

```python
import math

# Constants
print(f'pi = {math.pi}')
print(f'e  = {math.e}')
print(f'tau = {math.tau}')

# Rounding
print(math.floor(3.9))  # 3 — round down
print(math.ceil(3.1))   # 4 — round up
print(round(3.5))       # 4 — built-in round

# Powers and logs
print(math.pow(2, 10))  # 1024.0
print(math.log(100, 10)) # 2.0 — log base 10
print(math.log2(8))     # 3.0

# Trigonometry
print(math.sin(math.pi/2))  # 1.0
print(math.degrees(math.pi)) # 180.0
```

## 3. The `random` Module

```python
import random

# Random float between 0 and 1
print(random.random())

# Random int in range [a, b] (inclusive)
print(random.randint(1, 6))  # simulates a die roll

# Random choice from a sequence
fruits = ['apple', 'banana', 'cherry', 'date']
print(random.choice(fruits))

# Random sample (multiple unique choices)
print(random.sample(fruits, 2))

# Shuffle a list in place
deck = list(range(1, 6))
random.shuffle(deck)
print(deck)

# Seed for reproducible results
random.seed(42)
print([random.randint(1, 100) for _ in range(5)])
```

## 4. The `datetime` Module

```python
from datetime import datetime, date, timedelta

# Current date and time
now = datetime.now()
print(f'Now: {now}')
print(f'Date: {now.date()}')
print(f'Time: {now.time()}')
print(f'Year: {now.year}, Month: {now.month}, Day: {now.day}')

# Formatting
print(now.strftime('%Y-%m-%d'))           # '2024-01-15'
print(now.strftime('%d/%m/%Y %H:%M'))     # '15/01/2024 14:30'
print(now.strftime('%A, %B %d, %Y'))     # 'Monday, January 15, 2024'

# Date arithmetic
tomorrow = now + timedelta(days=1)
last_week = now - timedelta(weeks=1)
print(f'Tomorrow: {tomorrow.date()}')
print(f'Last week: {last_week.date()}')

# Days between two dates
start = date(2024, 1, 1)
end = date(2024, 12, 31)
diff = end - start
print(f'Days in 2024: {diff.days}')
```

## 6. The `json` Module

```python
import json

# Python dict → JSON string
person = {'name': 'Alice', 'age': 30, 'hobbies': ['reading', 'coding']}
json_string = json.dumps(person, indent=2)
print(json_string)

# JSON string → Python dict
parsed = json.loads(json_string)
print(type(parsed))      # <class 'dict'>
print(parsed['name'])    # Alice

# Write to file
with open('/tmp/person.json', 'w') as f:
    json.dump(person, f, indent=2)

# Read from file
with open('/tmp/person.json', 'r') as f:
    loaded = json.load(f)
print('Loaded:', loaded)
```

## 8. The `__name__ == '__main__'` Guard

When Python runs a file directly, `__name__` is set to `'__main__'`.  
When the file is **imported**, `__name__` is set to the module's name.

Use this guard to run code only when executing the file directly, not when importing.

```python
# Simulate what a module file looks like

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# This block runs only when you execute this file directly
# NOT when someone imports it
if __name__ == '__main__':
    print('Running as main script')
    print(add(3, 4))
    print(multiply(3, 4))

# Demonstrate __name__
print(f'__name__ in this notebook cell: {__name__}')
```

## 9. The `collections` Module — Specialized Containers

```python
from collections import Counter, defaultdict, OrderedDict, namedtuple

# Counter — count occurrences
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
counter = Counter(words)
print('Counter:', counter)
print('Most common:', counter.most_common(2))

# defaultdict — dict with a default value for missing keys
dd = defaultdict(list)
dd['fruits'].append('apple')
dd['fruits'].append('banana')
dd['veggies'].append('carrot')
print('defaultdict:', dict(dd))

# namedtuple — tuple with named fields
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(f'Point: x={p.x}, y={p.y}')
```

## Common Mistakes

### Mistake 1: Circular Imports

```python
# BAD — avoid circular imports (module_a imports module_b, module_b imports module_a)
# This causes ImportError.

# Solution: restructure your code so the shared logic is in a third module
# that both import from.

print('Circular imports cause ImportError — restructure to avoid them')
```

### Mistake 2: `from module import *` Pollutes Namespace

```python
# BAD
# from os import *    # imports everything — may overwrite your own names
# from math import *  # now 'log', 'floor' etc. could conflict

# GOOD — be explicit
from os.path import join, exists, basename
from math import sqrt, pi
print(join('a', 'b', 'c'))
print(sqrt(9))
```

### Exercise 2 (Medium): Word Frequency Counter
Use `collections.Counter` to find the 10 most common words in a paragraph of text. Ignore short words (length < 4).

```python
text = """
Python is an interpreted high-level general-purpose programming language.
Python's design philosophy emphasizes code readability. Its language constructs
and object-oriented approach aim to help programmers write clear, logical code
for small and large-scale projects. Python is dynamically typed and garbage-collected.
"""
# Your solution here
```

```python
# Solution
from collections import Counter
import re

words = re.findall(r'\b[a-z]+\b', text.lower())
filtered = [w for w in words if len(w) >= 4]
counter = Counter(filtered)

print('Top 10 words (length >= 4):')
for word, count in counter.most_common(10):
    print(f'  {word:15} {count}')
```

### Exercise 3 (Hard): Config File Reader
Write a function that reads a JSON config file and provides default values for missing keys.

```python
# Your solution here
```

```python
# Solution
import json
import os

DEFAULTS = {
    'host': 'localhost',
    'port': 8080,
    'debug': False,
    'log_level': 'INFO',
    'max_connections': 100
}

def load_config(filepath):
    config = dict(DEFAULTS)  # start with defaults
    if os.path.exists(filepath):
        with open(filepath) as f:
            user_config = json.load(f)
        config.update(user_config)  # override with user values
    return config

# Test with a config written to temp
test_config = {'port': 9000, 'debug': True}
with open('/tmp/test_config.json', 'w') as f:
    json.dump(test_config, f)

config = load_config('/tmp/test_config.json')
for key, value in config.items():
    print(f'{key}: {value}')
```

## Mini-Project: File Organizer

Organize files in a directory by extension using `os` and `shutil`.

```python
import os
import shutil

EXTENSION_MAP = {
    'Images':     ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.svg'],
    'Documents':  ['.pdf', '.doc', '.docx', '.txt', '.md'],
    'Videos':     ['.mp4', '.avi', '.mov', '.mkv'],
    'Audio':      ['.mp3', '.wav', '.flac', '.aac'],
    'Code':       ['.py', '.js', '.html', '.css', '.java'],
    'Data':       ['.csv', '.json', '.xml', '.xlsx'],
    'Archives':   ['.zip', '.tar', '.gz', '.rar'],
}

def get_category(filename):
    """Return the category for a filename based on its extension."""
    ext = os.path.splitext(filename)[1].lower()
    for category, extensions in EXTENSION_MAP.items():
        if ext in extensions:
            return category
    return 'Other'

def organize_directory(source_dir, dry_run=True):
    """
    Move files into categorized subdirectories.
    dry_run=True: only print what would happen.
    """
    if not os.path.exists(source_dir):
        print(f'Directory not found: {source_dir}')
        return
    
    moved = 0
    for filename in os.listdir(source_dir):
        filepath = os.path.join(source_dir, filename)
        if os.path.isfile(filepath):  # skip subdirectories
            category = get_category(filename)
            dest_dir = os.path.join(source_dir, category)
            dest_path = os.path.join(dest_dir, filename)
            
            print(f'  [{category}] {filename}')
            if not dry_run:
                os.makedirs(dest_dir, exist_ok=True)
                shutil.move(filepath, dest_path)
            moved += 1
    
    mode = 'DRY RUN' if dry_run else 'MOVED'
    print(f'\n{mode}: {moved} files would be organized')

# Demo with the current directory
print('Organizing current directory (dry run):')
organize_directory('.', dry_run=True)
```
