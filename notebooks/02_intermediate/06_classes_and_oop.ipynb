# Classes and Object-Oriented Programming (OOP)

**OOP** is a programming paradigm where you model the world as **objects** — things that have **attributes** (data) and **methods** (behavior).

| OOP Concept | Meaning |
|-------------|--------|
| **Class** | Blueprint/template for creating objects |
| **Object/Instance** | A specific thing created from a class |
| **Attribute** | Data stored in an object (like variables) |
| **Method** | Function that belongs to a class |
| **Encapsulation** | Bundling data + methods together |

By the end of this notebook you will be able to:
- Define classes with `__init__`, attributes, and methods
- Create and use objects
- Understand `self`
- Use class attributes vs instance attributes
- Implement `__str__` and `__repr__`

## 1. Your First Class

```python
class Dog:
    """Represents a dog."""
    
    # __init__ is called automatically when you create an object
    def __init__(self, name, breed, age):
        # 'self' refers to the object being created
        self.name = name    # instance attributes
        self.breed = breed
        self.age = age
    
    def bark(self):
        print(f'{self.name} says: Woof!')
    
    def describe(self):
        print(f'{self.name} is a {self.age}-year-old {self.breed}')


# Create objects (instances)
dog1 = Dog('Rex', 'German Shepherd', 3)
dog2 = Dog('Bella', 'Labrador', 5)

# Call methods
dog1.bark()
dog2.describe()

# Access attributes
print(dog1.name)
print(dog2.age)
```

## 2. Understanding `self`

`self` is how an object refers to **itself**. When you call `dog1.bark()`, Python automatically passes `dog1` as the first argument (`self`).

```python
class Counter:
    def __init__(self, start=0):
        self.count = start  # each Counter has its own 'count'
    
    def increment(self):
        self.count += 1
    
    def reset(self):
        self.count = 0
    
    def value(self):
        return self.count

c1 = Counter()
c2 = Counter(100)  # start at 100

c1.increment()
c1.increment()
c2.increment()

print(f'c1: {c1.value()}')  # 2 — c1 and c2 are independent
print(f'c2: {c2.value()}')  # 101
```

## 3. Class Attributes vs Instance Attributes

- **Instance attributes**: unique to each object (set via `self.attr = ...`)
- **Class attributes**: shared by ALL instances (defined directly in the class body)

```python
class Student:
    # Class attribute — shared by all students
    school = 'Python Academy'
    student_count = 0
    
    def __init__(self, name, grade):
        # Instance attributes — unique to each student
        self.name = name
        self.grade = grade
        Student.student_count += 1  # update class attribute
    
    def info(self):
        print(f'{self.name} | Grade: {self.grade} | School: {Student.school}')


s1 = Student('Alice', 'A')
s2 = Student('Bob', 'B')
s3 = Student('Charlie', 'A')

s1.info()
s2.info()

print(f'Total students: {Student.student_count}')  # 3

# Change class attribute — affects all instances
Student.school = 'Python University'
s1.info()  # now shows 'Python University'
```

## 4. `__str__` and `__repr__`

- `__str__`: human-readable string (used by `print()`)
- `__repr__`: developer-oriented string (used in the REPL)

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        return f'"{self.title}" by {self.author}'
    
    def __repr__(self):
        return f'Book(title={self.title!r}, author={self.author!r}, pages={self.pages})'

book = Book('Python Crash Course', 'Eric Matthes', 544)

print(book)        # uses __str__: '"Python Crash Course" by Eric Matthes'
print(repr(book))  # uses __repr__: Book(title='...', ...)
print([book])      # lists use __repr__
```

## 5. Methods: Instance, Class, and Static

| Type | Decorator | First Param | Access |
|------|-----------|-------------|--------|
| Instance method | (none) | `self` | Instance + class data |
| Class method | `@classmethod` | `cls` | Class data only |
| Static method | `@staticmethod` | (none) | Neither |

```python
class Temperature:
    unit = 'Celsius'  # class attribute
    
    def __init__(self, degrees):
        self.degrees = degrees
    
    # Instance method — works with self
    def to_fahrenheit(self):
        return self.degrees * 9/5 + 32
    
    # Class method — alternate constructor pattern
    @classmethod
    def from_fahrenheit(cls, fahrenheit):
        celsius = (fahrenheit - 32) * 5/9
        return cls(celsius)
    
    # Static method — utility, doesn't need self or cls
    @staticmethod
    def is_freezing(celsius):
        return celsius <= 0
    
    def __str__(self):
        return f'{self.degrees:.1f}°{Temperature.unit}'

t1 = Temperature(100)  # 100°C
print(t1)
print(f'In Fahrenheit: {t1.to_fahrenheit()}')

t2 = Temperature.from_fahrenheit(32)  # 0°C
print(t2)

print(Temperature.is_freezing(-5))   # True
print(Temperature.is_freezing(25))   # False
```

## 6. Properties — Controlled Attribute Access

`@property` lets you use method calls that look like attribute access.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius  # _ means 'private by convention'
    
    @property
    def radius(self):
        return self._radius
    
    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError('Radius cannot be negative')
        self._radius = value
    
    @property
    def area(self):  # computed property — no setter needed
        import math
        return math.pi * self._radius ** 2
    
    @property
    def circumference(self):
        import math
        return 2 * math.pi * self._radius

c = Circle(5)
print(f'Radius: {c.radius}')
print(f'Area: {c.area:.2f}')
print(f'Circumference: {c.circumference:.2f}')

c.radius = 10  # uses the setter
print(f'New area: {c.area:.2f}')

try:
    c.radius = -1
except ValueError as e:
    print(f'Error: {e}')
```

## Common Mistakes

### Mistake 1: Forgetting `self` in Method Definitions

```python
class BadExample:
    def __init__(self):
        self.value = 10
    
    # BAD — missing self
    def show():   # Python won't automatically pass the instance
        pass

class GoodExample:
    def __init__(self):
        self.value = 10
    
    # GOOD
    def show(self):
        print(self.value)

g = GoodExample()
g.show()
```

### Mistake 2: Mutable Default in `__init__`

```python
class BadCart:
    def __init__(self, items=[]):  # BAD — shared across all instances!
        self.items = items

class GoodCart:
    def __init__(self, items=None):  # GOOD
        self.items = items if items is not None else []
    
    def add(self, item):
        self.items.append(item)

cart1 = GoodCart()
cart2 = GoodCart()
cart1.add('apple')

print('cart1:', cart1.items)  # ['apple']
print('cart2:', cart2.items)  # [] — independent!
```

### Exercise 1 (Easy): Rectangle Class
Create a `Rectangle` class with width and height. Add methods for area and perimeter.

```python
# Solution
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def __str__(self):
        return f'Rectangle({self.width}x{self.height})'

r = Rectangle(5, 3)
print(r)
print(f'Area: {r.area()}')
print(f'Perimeter: {r.perimeter()}')
```

### Exercise 2 (Medium): Stack Data Structure
Implement a `Stack` class (last-in, first-out). Methods: `push`, `pop`, `peek`, `is_empty`, `size`.

```python
# Your solution here
```

```python
# Solution
class Stack:
    def __init__(self):
        self._items = []
    
    def push(self, item):
        self._items.append(item)
    
    def pop(self):
        if self.is_empty():
            raise IndexError('pop from empty stack')
        return self._items.pop()
    
    def peek(self):
        if self.is_empty():
            raise IndexError('peek from empty stack')
        return self._items[-1]
    
    def is_empty(self):
        return len(self._items) == 0
    
    def size(self):
        return len(self._items)
    
    def __str__(self):
        return f'Stack({self._items})'

s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s)
print('Peek:', s.peek())
print('Pop:', s.pop())
print(s)
```

Create a `BankAccount` class with deposit, withdraw, and transaction history.

```python
# Solution
from datetime import datetime

class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner = owner
        self._balance = initial_balance
        self._transactions = []
        if initial_balance > 0:
            self._log('Initial deposit', initial_balance)
    
    def _log(self, desc, amount):
        self._transactions.append({
            'time': datetime.now().strftime('%H:%M:%S'),
            'description': desc,
            'amount': amount,
            'balance': self._balance
        })
    
    @property
    def balance(self):
        return self._balance
    
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('Deposit must be positive')
        self._balance += amount
        self._log(f'Deposit', amount)
        print(f'Deposited £{amount:.2f}. Balance: £{self._balance:.2f}')
    
    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError('Withdrawal must be positive')
        if amount > self._balance:
            raise ValueError(f'Insufficient funds. Balance: £{self._balance:.2f}')
        self._balance -= amount
        self._log(f'Withdrawal', -amount)
        print(f'Withdrew £{amount:.2f}. Balance: £{self._balance:.2f}')
    
    def statement(self):
        print(f'\n--- Statement for {self.owner} ---')
        for t in self._transactions:
            sign = '+' if t['amount'] > 0 else ''
            print(f"{t['time']}  {t['description']:20} {sign}{t['amount']:8.2f}  Balance: £{t['balance']:.2f}")
        print(f'Current Balance: £{self._balance:.2f}')

acc = BankAccount('Alice', 1000)
acc.deposit(500)
acc.withdraw(200)
acc.deposit(100)
try:
    acc.withdraw(5000)
except ValueError as e:
    print(f'Error: {e}')
acc.statement()
```
